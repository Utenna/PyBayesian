### How to choose an appropriate likelihood


å‡è®¾ä½ ç°åœ¨æ˜¯åŒ»é™¢åŒ»ç”Ÿï¼Œè¦ä¼°è®¡æŸç§ç–¾ç—…çš„æ‚£ç—…ç‡ã€‚æˆ‘ä»¬ä»ä¸€ä¸ªæ ·æœ¬å¼€å§‹ï¼Œæˆ‘ä»¬æ²¡æœ‰å…ˆéªŒä¿¡æ¯ã€‚è®©è¿™ä¸ªäººï¼ˆæ ·æœ¬ï¼‰çš„ç–¾ç—…çŠ¶æ€ç”¨$X$æ¥è¡¨ç¤º 

$$

$$


$\theta$ä»£è¡¨ä¸€ä¸ªéšæœºé€‰æ‹©çš„ä¸ªä½“æ‚£æœ‰è¿™ç§ç–¾ç—…çš„æ¦‚ç‡ã€‚æˆ‘ä»¬ç°åœ¨è®¡ç®—ä¸€ä¸ªæ ·æœ¬æ¯ä¸€ä¸ªå¯èƒ½ç»“æœçš„æ¦‚ç‡

$P(X=0|\theta)=(1-\theta)$

$P(X=1|\theta)=(\theta)$

æˆ‘ä»¬å¯ä»¥æ ¹æ®ä¸Šé¢çš„å…¬å¼ï¼ŒåŸºäº1æˆ–æ˜¯0ï¼ˆç”Ÿç—…è¿˜æ˜¯æ²¡ç”Ÿç—…ï¼‰å†™ä¸‹ä¸€ä¸ªè¡¨è¾¾å¼ï¼š

$$
P(X=a|\theta)=\theta^a(1-\theta)^{1-a} \qquad where \quad a\in \left\{0ï¼Œ 1\right\}
$$

æˆ‘ä»¬å¯ä»¥ä½¿ç”¨pythonä»£ç æ¥å®ç°ä¸Šé¢çš„ä¾‹å­ã€‚é¦–å…ˆå…ˆåŠ è½½numpyåº“ã€‚

ä¹‹åæˆ‘ä»¬å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œnå°±æ˜¯æ ·æœ¬é‡ï¼Œåœ¨æˆ‘ä»¬ä¾‹å­ä¸­ä¸º1ï¼›aæ˜¯æˆ‘ä»¬æ ·æœ¬ç”Ÿç—…è¿˜æ˜¯æ²¡ç”Ÿç—…çš„æƒ…å†µï¼›æœ€åæ˜¯$\theta$ï¼Œæˆ‘ä»¬éšæœºé€‰ä¸­ä¸€ä¸ªä¸ªä½“ç”Ÿç—…çš„æ¦‚ç‡ã€‚

$Z=X_1+X_2$

æ‚£ç—…äººæ•°çš„æƒ…å†µå¯ä»¥ç”¨ä¸‹åˆ—å…¬å¼è¿›è¡Œè¡¨ç¤ºï¼š

$Pr(Z=0|\theta) = Pr(X_1 =0 , X_2 = 0| \theta) = (1-\theta)^2$

$Pr(Z=1|\theta) = Pr(X_1 =1 , X_2 = 0| \theta) = 2\theta(1-\theta)$

$Pr(Z=2|\theta) = Pr(X_1 =1 , X_2 = 1| \theta) = \theta^2$

è¿™ä¸ªå…¬å¼å¯ä»¥ç”¨ä¸€ä¸ªæ™®éçš„å½¢å¼è¿›è¡Œè¡¨è¾¾ï¼š

$Pr(Z=\beta|\theta) = \binom{2}{\beta} \theta^{\beta}(1-\theta)^{2-\beta}$

#### å½“N=3æ—¶ï¼Œæ‚£ç—…äººæ•°çš„æƒ…å†µå¯ä»¥ç”¨ä¸‹åˆ—å…¬å¼è¿›è¡Œè¡¨ç¤ºï¼š


$Pr(Z=0|\theta) = Pr(X_1 =0 , X_2 = 0, X_3 = 0| \theta)$

$Pr(Z=1|\theta) = 3Pr(X_1 =1 , X_2 = 0, X_3 = 0| \theta)$

$Pr(Z=2|\theta) = 3Pr(X_1 =1 , X_2 = 1, X_3 = 0| \theta)$

$Pr(Z=3|\theta) = Pr(X_1 =1 , X_2 = 1, X_3 = 1| \theta)$

è¿™ä¸ªå…¬å¼å¯ä»¥ç”¨ä¸€ä¸ªæ™®éçš„å½¢å¼è¿›è¡Œè¡¨è¾¾ï¼š

$Pr(Z=\beta|\theta) = \binom{3}{\beta} \theta^{\beta}(1-\theta)^{3-\beta}$

#### Example:

è‹¥$\theta_1=0.5$ï¼Œ3ä¸ªäººä¸­æœ‰ä¸€ä¸ªäººæ‚£ç—…çš„æ¦‚ç‡ä¸ºå¤šå°‘ï¼Ÿ

```python
# combè¿™ä¸ªå‡½æ•°æ˜¯ä¸“é—¨ç”¨æ¥ç®—ç»„åˆæ•°çš„
from scipy.special import comb
```

```python
# å®šä¹‰ä¸€ä¸ªå‡½æ•°ï¼Œè¿™ä¸ªå‡½æ•°å¯ä»¥ç”¨æ¥è®¡ç®—nä¸ªä¸ªä½“ä¸­mäººæ‚£ç—…çš„æ¦‚ç‡

# def å®šä¹‰å‡½æ•°åï¼Œ()å†…ä¸ºå‡½æ•°çš„å‚æ•°
# nä¸ºä¸ªä½“æ•°
# mä¸ºæ‚£ç—…äººæ•°
# thetaä¸ºä¸ªä½“æ‚£ç—…çš„æ¦‚ç‡
def likelihood(n,m,theta):
    return comb(n,m)*((theta)**(m))*((1-theta)**(n-m))

# ä¸ªä½“æ‚£ç—…æ¦‚ç‡ä¸º0.8
theta = 0.8
# å…±æœ‰3ä¸ªä¸ªä½“
n=3
# æ‚£ç—…äººæ•°ä¸º1
m=1
# è®¡ç®—ä¸ªä½“æ‚£ç—…æ¦‚ç‡ä¸º0.8çš„æƒ…å†µä¸‹ï¼Œ3ä¸ªä¸ªä½“ä¸­1äººæ‚£ç—…çš„æ¦‚ç‡
likelihood = likelihood(n,m,theta)


# è¾“å‡ºç»“æœ
print(f"ä¸ªä½“æ‚£ç—…æ¦‚ç‡ä¸º0.8çš„æƒ…å†µä¸‹ï¼Œ3ä¸ªä¸ªä½“ä¸­1äººæ‚£ç—…çš„æ¦‚ç‡ä¸º{likelihood}")
```

# Bonus:ä¸€ç»„è¢«è¯•çš„æ™ºåŠ›æµ‹è¯•åˆ†æ•°

å‡è®¾æˆ‘ä»¬å¯¹ä¸€ç»„è¢«è¯•è¿›è¡Œæ™ºåŠ›æµ‹è¯•ï¼Œæµ‹è¯•åˆ†æ•°çš„èŒƒå›´æ˜¯0-200ã€‚å‡è®¾æˆ‘ä»¬æ²¡æœ‰å…³äºä¸ªä½“ç‰¹å¾çš„ä¿¡æ¯æ¥å¸®åŠ©é¢„æµ‹ï¼ŒåŒæ—¶ä¸€ä¸ªäººçš„æ™ºåŠ›æµ‹è¯•åˆ†æ•°å¯èƒ½ä¼šæœ‰å¤šä¸ªå› ç´ å½±å“ï¼Œæˆ‘ä»¬å‡è®¾ä¸€ä¸ªäººçš„åˆ†æ•°æ˜¯æ‰€æœ‰è¿™äº›å› ç´ çš„å¹³å‡å€¼ï¼Œé‚£ä¹ˆä¸­å¿ƒæé™å®šç†å¯èƒ½é€‚ç”¨ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨æ­£æ€åˆ†å¸ƒæ¥æ¨¡æ‹Ÿæµ‹è¯•åˆ†æ•°Xï¼š

$p(X=a|\mu, \sigma)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(a-\mu)^2}{2\sigma^2}}$

ç°åœ¨è€ƒè™‘ä¸€ä¸ªç”±Nä¸ªä¸ªä½“ç»„æˆçš„æ ·æœ¬ã€‚å¦‚æœæˆ‘ä»¬å‡è®¾ä¸€ä¸ªäººçš„è€ƒè¯•æˆç»©æ— åŠ©äºé¢„æµ‹å¦ä¸€ä¸ªäººçš„æˆç»©ï¼Œé‚£ä¹ˆæˆ‘ä»¬çš„æ•°æ®å°±æ˜¯ç‹¬ç«‹çš„ã€‚å¦‚æœæˆ‘ä»¬è¿˜å‡è®¾æ‰€æœ‰çš„ä¸ªä½“éƒ½æ¥è‡ªåŒä¸€ä¸ªç¾¤ä½“ï¼Œæˆ‘ä»¬å°±æœ‰ä¸€ä¸ªéšæœºæ ·æœ¬ã€‚ç„¶åï¼Œæˆ‘ä»¬å¯ä»¥é€šè¿‡å–ä¸ªä½“æ¦‚ç‡å¯†åº¦çš„ä¹˜ç§¯æ¥è®¡ç®—Nä¸ªä¸ªä½“æ ·æœ¬çš„è”åˆæ¦‚ç‡å¯†åº¦ï¼š

$p(X_1=a_1, X_2=a_2, ..., X_N=a_N|\mu, \sigma)=\prod\limits_{i=1}^N\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(a-\mu)^2}{2\sigma^2}}$

```python
# å®šä¹‰æ­£æ€åˆ†å¸ƒçš„æ¦‚ç‡å¯†åº¦å‡½æ•°ï¼Œaä¸ºæ•°æ®ï¼Œmuä¸ºå¹³å‡å€¼ï¼Œsigmaä¸ºæ ‡å‡†å·®
def pdf(a, mu, sigma):

    # æ­£æ€åˆ†å¸ƒçš„æ¦‚ç‡å¯†åº¦å‡½æ•°
    fx = np.exp(-(a-mu)**2/(2*sigma**2))*1/np.sqrt(2*np.pi*sigma**2)
    
    # è¿”å›æ­£æ€åˆ†å¸ƒçš„æ¦‚ç‡å¯†åº¦å‡½æ•°çš„å€¼
    return fx

# ä»¥60ä¸ºèµ·ç‚¹ï¼Œ180ä¸ºç»ˆç‚¹(ä¸åŒ…æ‹¬80)ï¼Œæ­¥é•¿ä¸º20ç”Ÿæˆæ•°ç»„
a_n = np.arange(60, 180, 20)

#åœ¨pdfå‡½æ•°ä¸­ï¼Œmuæ˜¯æ€»ä½“å‡å€¼ï¼Œ100æ˜¯æ€»ä½“æ ‡å‡†å·®ï¼Œaæ˜¯ä¸ªä½“çš„æ™ºåŠ›æµ‹éªŒåˆ†æ•°
a_pdf = pdf(a_n, 80, 100)

#np.prodæ˜¯è¿ä¹˜å‡½æ•°
likelihood = np.prod(a_pdf)

print(f"è§‚å¯Ÿåˆ°è¿™ç»„æ•°æ®çš„æ¦‚ç‡ä¸º{likelihood}")
```

### 3.2 Maximum likelihood çš„çœŸå®è®¡ç®—è¿‡ç¨‹

- è·å–å®éªŒæ•°æ®ï¼Œæ¯”å¦‚ï¼Œ100ä¸­æœ‰10äººæ‚£ç—…ã€‚

- é€šè¿‡å¯¹æ•°logå°†ä¼¼ç„¶å‡½æ•°ä¸­çš„è¿ä¹˜å˜ä¸ºæ±‚å’Œï¼Œæ¯”å¦‚ $log(\theta^{\beta}(1-\theta)^{n-\beta}) = log(\theta^{\beta}) + log((1-\theta)^{n-\beta})$, $log(0.1^{10}(1-0.1)^{100-10}) = log(0.1^{10}) + log((1-0.1)^{100-10}) = -23.03 + -9.48 = -32.51$ã€‚
