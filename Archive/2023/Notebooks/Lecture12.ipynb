### Linear regression model  

é¦–å…ˆï¼Œæˆ‘ä»¬ä½¿ç”¨ä¸ŠèŠ‚è¯¾å­¦ä¹ è¿‡çš„æ­£æ€å›å½’æ¨¡å‹æ¥è¿›è¡Œå»ºæ¨¡åˆ†æï¼Œè™½ç„¶æˆ‘ä»¬çŸ¥é“è®¡æ•°æ•°æ®å¯èƒ½ä¸é€‚åˆæ­£æ€åˆ†å¸ƒï¼Œä½†æ˜¯ï¼Œæ­£æ€åˆ†å¸ƒçš„æ•°æ®å°†èƒ½æä¾›ä¸€äº›ç›´è§‰ï¼Œå‘Šè¯‰æˆ‘ä»¬ï¼Œæ¨¡å‹é”™åœ¨å“ªé‡Œã€‚  

$$  
 
\begin{array}{lcrl}  

\text{data:} & \hspace{.05in} & Y_i|\beta_0,\beta_1,\beta_2,\sigma & \stackrel{ind}{\sim} N(\mu_i, \; \sigma^2)  \;\; \text{ with } \;\; \mu_i = \beta_0 + \beta_1X_{i1} + \beta_2X_{i2} \\  
\text{priors:} & & \beta_{0}  & \sim N\left(20, 10^2 \right)  \\  
                    & & \beta_1  & \sim N\left(0, 10^2 \right) \\  
                    & & \beta_2  & \sim N\left(0, 10^2 \right) \\  
                    & & \sigma   & \sim \text{Exp}(0.6)  .\\  
\end{array}  

$$  

* æˆ‘ä»¬å¿«é€Ÿè¿è¡Œæ¨¡å‹å®šä¹‰ã€MCMCé‡‡æ ·å’ŒåéªŒé¢„æµ‹åˆ†å¸ƒçš„ç»“æœ

```python
with pm.Model(coords = {"obs_id": df.index}) as normal_model:

    beta_0 = pm.Normal("beta_0", mu=20, sigma=10)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=10)           #å®šä¹‰beta_1
    beta_2 = pm.Normal("beta_2", mu=0, sigma=10)           #å®šä¹‰beta_2
    sigma = pm.Exponential("sigma", 0.6)                   #å®šä¹‰sigma

    stress = pm.MutableData("stress",df.stress, dims="obs_id")                           #è‡ªå˜é‡1:å‹åŠ›æ°´å¹³
    romantic = pm.MutableData("romantic",df.romantic, dims="obs_id")                     #è‡ªå˜é‡2:æ‹çˆ±æƒ…å†µ
    mu = pm.Deterministic("mu", beta_0 + beta_1*stress + beta_2*romantic, dims="obs_id") #å®šä¹‰muï¼Œè‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ

    likelihood = pm.Normal("y_est", mu=mu, sigma=sigma, observed=df.networksize, dims="obs_id") 
```

```python
# å¯è§†åŒ–æ¨¡å‹ï¼ˆè´å¶æ–¯å˜é‡å› æœå›¾ï¼‰ã€‚ 
# æˆ‘ä»¬ä½¿ç”¨ pymc è‡ªå¸¦çš„ `model_to_graphviz` æ–¹æ³•æ¥å¯è§†åŒ–æ¨¡å‹ä¸­å„å˜é‡çš„å› æœå…³ç³»ã€‚
pm.model_to_graphviz(normal_model)
```

```python
#===========================
#     æ³¨æ„ï¼ï¼ï¼ä»¥ä¸‹ä»£ç å¯èƒ½éœ€è¦è¿è¡Œ1-2åˆ†é’Ÿå·¦å³
#===========================
with normal_model:
    # MCMC è¿‘ä¼¼åéªŒåˆ†å¸ƒ
    normal_model_trace = pm.sample(
        draws=5000,                   # ä½¿ç”¨mcmcæ–¹æ³•è¿›è¡Œé‡‡æ ·ï¼Œdrawsä¸ºé‡‡æ ·æ¬¡æ•°
        tune=1000,                    # tuneä¸ºè°ƒæ•´é‡‡æ ·ç­–ç•¥çš„æ¬¡æ•°ï¼Œå¯ä»¥å†³å®šè¿™äº›ç»“æœæ˜¯å¦è¦è¢«ä¿ç•™
        chains=4,                     # é“¾æ•°
        discard_tuned_samples= True,  # tuneçš„ç»“æœå°†åœ¨é‡‡æ ·ç»“æŸåè¢«ä¸¢å¼ƒ
        random_seed=84735)
    
    # åéªŒé¢„æµ‹
    normal_model_ppc = pm.sample_posterior_predictive(normal_model_trace, random_seed=84735)
```

```python
# å®šä¹‰è®¡ç®— MAE å‡½æ•°
from statistics import median
def MAE(model_ppc):
    # è®¡ç®—æ¯ä¸ªXå–å€¼ä¸‹å¯¹åº”çš„åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼
    pre_y = model_ppc.posterior_predictive["y_est"].stack(sample=("chain", "draw"))
    pre_y_mean = pre_y.mean(axis=1).values

    # æå–è§‚æµ‹å€¼Yï¼Œæå–å¯¹åº”Yå€¼ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼
    MAE = pd.DataFrame({
        "networksize_ppc_mean": pre_y_mean,
        "networksize_original": df.networksize
    })

    # è®¡ç®—é¢„æµ‹è¯¯å·®
    MAE["pre_error"] = abs(MAE["networksize_original"] -\
                            MAE["networksize_ppc_mean"])

    # æœ€åï¼Œè®¡ç®—é¢„æµ‹è¯¯å·®çš„ä¸­ä½æ•°
    MAE = median(MAE.pre_error)
    return MAE
```

```python
# å®šä¹‰è®¡ç®—è¶…å‡ºHDIçš„æ•°é‡
def counter_outlier(model_ppc, hdi_prob=0.95):
    # å°†az.summaryç”Ÿæˆçš„ç»“æœå­˜åˆ°hdi_multiè¿™ä¸ªå˜é‡ä¸­ï¼Œè¯¥å˜é‡ä¸ºæ•°æ®æ¡†
    hdi = az.summary(model_ppc, kind="stats", hdi_prob=hdi_prob)
    lower = hdi.iloc[:,2].values
    upper = hdi.iloc[:,3].values

    # å°†åŸæ•°æ®ä¸­çš„è‡ªæˆ‘æ§åˆ¶åˆ†æ•°åˆå¹¶ï¼Œä¾¿äºåç»­è¿›è¡Œåˆ¤æ–­
    y_obs = model_ppc.observed_data["y_est"].values

    # åˆ¤æ–­åŸæ•°æ®ä¸­çš„å‹åŠ›åˆ†æ•°æ˜¯å¦åœ¨åéªŒé¢„æµ‹çš„95%å¯ä¿¡åŒºé—´å†…ï¼Œå¹¶è®¡æ•°
    hdi["verify"] = (y_obs <= lower) | (y_obs >= upper)
    hdi["y_obs"] = y_obs
    hdi_num = sum(hdi["verify"])

    return hdi_num
```

```python
normal_model_MAE = MAE(normal_model_ppc)
print(f"æ­£æ€å›å½’æ¨¡å‹MAE: {normal_model_MAE:.2f}")

normal_model_outliers = counter_outlier(normal_model_ppc)
print(f"æ­£æ€å›å½’æ¨¡å‹è¶…å‡º95%hdi: {normal_model_outliers:.2f}")
```

## Poisson regression  

> åœ¨lec5ä¸­ï¼Œæˆ‘ä»¬æ›¾ä»‹ç»è¿‡æ³Šæ¾åˆ†å¸ƒï¼Œå½“æ—¶ä½¿ç”¨çš„ä¾‹å­ä¸ºåœ¨äºšè¿ä¼šæ—¶ï¼Œä¸­å›½æ¯å¤©è·å¾—çš„é‡‘ç‰Œæ•°é‡ã€‚  

* **æ³Šæ¾åˆ†å¸ƒ(Poisson distribution)** è¡¨ç¤ºåœ¨ç»™å®š$\lambda$ä¸‹ï¼Œç”¨æ¥æè¿°æ—¶é—´åœ¨ä¸€å®šçš„æ—¶é—´æˆ–ç©ºé—´å†…å‘ç”Ÿçš„äº‹ä»¶ä¸ªæ•°ï¼Œ$\lambda$ä¸ºäº‹ä»¶çš„å¹³å‡å‘ç”Ÿç‡æˆ–äº‹ä»¶å‘ç”Ÿçš„å‡å€¼  

![Image Name](https://cdn.kesci.com/upload/s4o80dsjlv.png?imageView2/0/w/960/h/960)  



* åœ¨æ­£æ€å›å½’æ¨¡å‹ä¸­ï¼Œå¯¹æ•°æ®çš„å‡è®¾ä¸º $Y_i|\beta_0,\beta_1,\beta_2,\sigma  \stackrel{ind}{\sim} N(\mu_i, \; \sigma^2)  \;\; \text{ with } \;\; \mu_i = \beta_0 + \beta_1X_{i1} + \beta_2X_{i2}$  

    > å³å› å˜é‡çš„å‡å€¼å¯ä»¥è¡¨ç¤ºä¸ºè‡ªå˜é‡çš„çº¿æ€§ç»„åˆï¼Œä¸”å› å˜é‡ç¬¦åˆä»¥è¯¥å‡å€¼ä¸ºä¸­å¿ƒçš„æ­£æ€åˆ†å¸ƒ <br>  
    > å…¶ä¸­ X1 ä¸ºå‹åŠ›ï¼ŒX2 ä¸ºæ‹çˆ±æƒ…å†µ(1ä»£è¡¨æ‹çˆ±ï¼Œ0ä»£è¡¨æ²¡æœ‰æ‹çˆ±)  


* ç±»ä¼¼çš„ï¼Œæˆ‘ä»¬ä¹Ÿå¯ä»¥åœ¨æ³Šæ¾å›å½’æ¨¡å‹ä¸­åšå‡ºè¿™æ ·çš„å‡è®¾ï¼š  
$Y_i | \lambda_i \stackrel{ind}{\sim} Pois\left(\lambda_i \right) \;\; \text{ with } \;\; \lambda_i = \beta_0 + \beta_1X_{i1} + \beta_2X_{i2}$  

**éœ€è¦æ³¨æ„çš„æ˜¯ï¼Œåœ¨å…¬å¼(1)ä¸­ï¼Œäº‹ä»¶å‘ç”Ÿæ•°é‡æˆ–äº‹ä»¶å‘ç”Ÿç‡ $\lambda_i$ çš„èŒƒå›´ä¸º 0~âˆä¹‹é—´ï¼Œç„¶è€Œ $\beta_0 + \beta_1X_{i1} + \beta_2X_{i2}$ çš„é¢„æµ‹å€¼ä¼šå‡ºç°è´Ÿæ•°ï¼Œè¿™æ˜¾ç„¶æ˜¯ä¸åˆç†çš„**ã€‚

### æ³Šæ¾å›å½’æ¨¡å‹å…¬å¼  

ä¸ºäº†è§£å†³è¿™ä¸ªé—®é¢˜ï¼Œæˆ‘ä»¬å¯ä»¥å¯¹ $\lambda_i$ è¿›è¡Œå¯¹æ•°è½¬æ¢  

* æˆ‘ä»¬å‡è®¾ $\lambda_i$ çš„èŒƒå›´ä¸º 0~50ã€‚  
* ç»è¿‡å¯¹æ•°è½¬åŒ–ï¼Œ$\text{log}(\lambda_i)$(çºµåæ ‡)çš„èŒƒå›´å˜ä¸º -4~4ã€‚

```python
# ç”Ÿæˆ lambdas çš„èŒƒå›´ä¸º 0åˆ°50
lambdas = np.arange(0.01,50,0.2)
# è¿›è¡Œå¯¹æ•°è½¬æ¢
log_lambdas = np.log(lambdas)

plt.plot(lambdas, log_lambdas)
plt.xlabel("lambdas")
plt.ylabel("log_lambdas")
plt.show()
```

* æ­¤æ—¶ï¼Œ$log(\lambda_i)$æ—¢å¯ä»¥æ˜¯æ­£æ•°ä¹Ÿå¯ä»¥æ˜¯è´Ÿæ•°  
$$  
Y_i | \beta_0,\beta_1, \beta_2 \stackrel{ind}{\sim} Pois\left(\lambda_i \right)  
\;\;\; \text{ with } \;\;\;  
\log\left( \lambda_i \right) = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}.  
$$  

* å› æ­¤ï¼Œæ³Šæ¾å›å½’æ¨¡å‹å¯ä»¥å†™æˆï¼š  
$$  
Y_i | \beta_0,\beta_1, \beta_2 \stackrel{ind}{\sim} Pois\left(\lambda_i\right) \;\; \text{ with } \;\; \lambda_i = e^{\beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}}  
$$  

> ğŸ“æ•°å­¦åŸºç¡€æé†’ï¼š  
> $$  
> \begin{equation}  
> \begin{split}  
> \log(\lambda) & = \beta_0 + \beta_1 X_1 + \cdots \beta_p X_p \\  
> \lambda & = e^{\beta_0 + \beta_1 X_1 + \cdots \beta_p X_p}. \\  
> \end{split}  
> \end{equation}  
> $$  


åœ¨è¿™é‡Œæˆ‘ä»¬æ¼”ç¤ºï¼Œåœ¨ç»™å®šçš„$x$ä¸‹ $\lambda_i = e^{\beta_0 + \beta_1 X_{i1}}$ å’Œ $\lambda_i = \beta_0 + \beta_1 X_{i1}$ çš„å·®å¼‚  
- å…¶ä¸­ï¼Œæˆ‘ä»¬å‡è®¾ x çš„èŒƒå›´ä¸º 0~20ï¼Œå¹¶ä¸” $beta_0=0$, $beta_1=0.2$ã€‚  
- ä½¿ç”¨ `np.exp` å¯ä»¥è¿›è¡Œé€†å¯¹æ•°è½¬æ¢ã€‚  
- å¯ä»¥å‘ç°ï¼Œé€†å¯¹æ•°è½¬æ¢çš„é¢„æµ‹ç»“æœæ˜¯**éçº¿æ€§**çš„ã€‚

### å…¬å¼ä¸­å„å‚æ•°çš„æ„ä¹‰  

$$  
Y_i | \beta_0,\beta_1, \beta_2 \stackrel{ind}{\sim} Pois\left(\lambda_i\right) \;\; \text{ with } \;\; \lambda_i = e^{\beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}}  
$$  

* $\beta_0$  

    * å½“$X_1= X_2 = 0$æ—¶ï¼Œ$\lambda_i = e^{\beta_0}$ï¼Œå½“è‡ªå˜é‡éƒ½ä¸º0æ—¶ï¼Œäº‹ä»¶çš„å¹³å‡å‘ç”Ÿç‡ä¸º $e^{\beta_0}$  

* $\beta_1, ..., \beta_p$  

    * å‡è®¾ $X_2= 0$ï¼Œ$X_1= 1$ æ—¶æœ‰ $\lambda_1 =e^{\beta_0 + \beta_1}$ï¼Œ$X_1= 2$ æ—¶æœ‰ $\lambda_2 =e^{\beta_0 + 2\beta_1}$  

    * æ­¤æ—¶æœ‰ $\frac{\lambda_2}{\lambda_1} = e^{\beta_1}$ ï¼Œ å› æ­¤å¯ä»¥å¾—åˆ° $\beta_1 = log(\lambda_2) - log(\lambda_1)$ æˆ–è€… $e^{\beta_1} = \frac{\lambda_{x+1}}{\lambda_x}$  
    
    > è¿™è¡¨æ˜ï¼šåœ¨å…¶ä»–è‡ªå˜é‡ä¸å˜çš„æƒ…å†µä¸‹ï¼Œè‡ªå˜é‡$X_1$æ¯å¢åŠ ä¸€ä¸ªå•ä½ï¼Œäº‹ä»¶çš„å¹³å‡å‘ç”Ÿç‡$\lambda_i$å°±ä¼šä»¥$e^{\beta_1}$å€å¢é•¿  


### æ³Šæ¾å›å½’çš„é¢„è®¾(assumptions)  

1. **è®¡æ•°æ•°æ®ï¼š** é€‚ç”¨äºå› å˜é‡æ˜¯è®¡æ•°æ•°æ®çš„æƒ…å†µï¼Œå³éè´Ÿæ•´æ•°ï¼Œä¾‹å¦‚äº‹ä»¶å‘ç”Ÿçš„æ¬¡æ•°ã€è®¡æ•°çš„æ•°é‡ç­‰ã€‚  

2. **ç‹¬ç«‹æ€§ï¼š** å‡è®¾æ¯ä¸ªè§‚æµ‹å€¼$Y_i$æ˜¯ç›¸äº’ç‹¬ç«‹çš„ï¼Œå³ä¸€ä¸ªè§‚æµ‹å€¼ä¸å—å…¶ä»–è§‚æµ‹å€¼çš„å½±å“ã€‚  

4. **çº¿æ€§å…³ç³»ï¼š** ç»å¯¹æ•°è½¬æ¢åçš„å› å˜é‡ å¯ä»¥å†™ä½œè‡ªå˜é‡çš„çº¿æ€§ç»„åˆï¼Œå¦‚ï¼š$\log\left( \lambda_i \right) = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}+ \beta_3 X_{i3}$  

3. **å‡å€¼ç­‰äºæ–¹å·®ï¼š** å¯¹äºæ³Šæ¾åˆ†å¸ƒ$Pois(\lambda_i)$æ¥è¯´ï¼Œå…¶å‡å€¼å’Œæ–¹å·®éƒ½ä¸äº‹ä»¶çš„å‘ç”Ÿç‡ç›¸ç­‰(æˆ–çº¦ç­‰)ï¼Œ$E(Y) = \text{Var}(Y) = \lambda$ã€‚  

    * éœ€è¦æ³¨æ„çš„æ˜¯ï¼Œè¿™å¹¶ä¸æ„å‘³ç€å› å˜é‡æœ¬èº«çš„åˆ†å¸ƒæ˜¯æ³Šæ¾åˆ†å¸ƒã€‚åœ¨æ³Šæ¾å›å½’ä¸­ï¼Œæˆ‘ä»¬è¯´çš„æ˜¯ï¼Œåœ¨ç»™å®šçš„Xä¸‹ï¼Œå…¶è§‚æµ‹åˆ°çš„Yéœ€è¦æ»¡è¶³æ³Šæ¾åˆ†å¸ƒ  

    * è‹¥æ­¤å‡è®¾è¢«æ»¡è¶³ï¼Œåœ¨æ•°æ®ä¸­æˆ‘ä»¬å¯ä»¥çœ‹åˆ°ï¼Œå› å˜é‡å‡å€¼å¢å¤§çš„åŒæ—¶ï¼Œå…¶å˜å¼‚æ€§ä¹Ÿåœ¨å¢å¤§  

    * è¯¥å‡è®¾æ˜¯æ³Šæ¾å›å½’çš„ä¸€ä¸ªé‡è¦å‡è®¾ï¼Œå½“å‡å€¼è¿œå¤§äºæ–¹å·®æ—¶ï¼Œæ³Šæ¾å›å½’å¯èƒ½å¹¶ä¸æ˜¯æœ€ä½³çš„æ¨¡å‹é€‰æ‹©ï¼Œä¹‹åæˆ‘ä»¬ä¼šä»‹ç»è¿™ç§æƒ…å†µä¸‹æ›´é€‚åˆçš„æ¨¡å‹é€‰æ‹©ã€‚

### æ¨¡å‹å®šä¹‰  

**å®šä¹‰æ³Šæ¾å›å½’æ¨¡å‹çš„å…ˆéªŒ**  

> æ³¨æ„ï¼šå®šä¹‰å…ˆéªŒæ—¶éœ€è¦é¢å¤–è€ƒè™‘å¯¹æ•°å˜åŒ–å¯¹å…ˆéªŒèŒƒå›´çš„å½±å“ã€‚  

$$  
\begin{equation}  
\begin{array}{lcrl}  
\text{data:} & \hspace{.025in} & Y_i|\beta_0,\beta_1,\beta_2 & \stackrel{ind}{\sim} Pois\left(\lambda_i \right) \;\; \text{ with } \;\;  
\lambda_i = e^{\beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}}\\  
\text{priors:} & & \beta_{0} & \sim N\left(0, 0.5^2 \right) \\  
&& \beta_1 & \sim N(0, 0.1^2) \\  
&& \beta_2 & \sim N(0, 0.5^2) \\  
\end{array}  
\end{equation}  
$$

```python
with pm.Model(coords = {"obs_id": df.index})  as poisson_model:

    beta_0 = pm.Normal("beta_0", mu=0, sigma=0.5)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=0.1)          #å®šä¹‰beta_1
    beta_2 = pm.Normal("beta_2", mu=0, sigma=0.5)          #å®šä¹‰beta_2

    #è‡ªå˜é‡1:å‹åŠ›æ°´å¹³
    stress = pm.MutableData("stress",df.stress, dims="obs_id")                        
    #è‡ªå˜é‡2:æ€§åˆ«
    romantic = pm.MutableData("romantic",df.romantic, dims="obs_id")            
    # é¢„æµ‹ lambdaï¼Œè‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ
    lam = pm.Deterministic(
        "lam", 
        ##------------------------------------------------
        #  æ³¨æ„ï¼Œè¿™é‡Œæˆ‘ä»¬ä½¿ç”¨ pm.math.exp å¯¹é¢„æµ‹è¿›è¡Œé€†å¯¹æ•°è½¬æ¢
        #------------------------------------------------
        pm.math.exp(beta_0 + beta_1*stress + beta_2*romantic), dims="obs_id")

    likelihood = pm.Poisson("y_est", mu=lam, observed=df.networksize, dims="obs_id") 
```

#### å…ˆéªŒé¢„æµ‹æ£€éªŒ  

* å‰é¢è¯´åˆ°ä½¿ç”¨æ­£æ€å›å½’æ¨¡å‹ä¼šå¯¼è‡´äº‹ä»¶çš„å‘ç”Ÿç‡$\lambda_i$ä¸ºè´Ÿæ•°ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨`pm.sample_prior_predictive`è¿›è¡Œå…ˆéªŒé¢„æµ‹æ£€éªŒï¼Œæ¥æŸ¥çœ‹ç”Ÿæˆçš„$\lambda_i$æƒ…å†µ

```python
# å¯¹æ­£æ€å›å½’æ¨¡å‹è¿›è¡Œå…ˆéªŒé¢„æµ‹
normal_prior =pm.sample_prior_predictive(samples=50, 
                                          model=normal_model,
                                          var_names=["mu"],
                                          random_seed=84735)
# å¯¹æ³Šæ¾å›å½’æ¨¡å‹è¿›è¡Œå…ˆéªŒé¢„æµ‹
poisson_prior =pm.sample_prior_predictive(samples=50, 
                                          model=poisson_model,
                                          var_names=["lam"],
                                          random_seed=84735)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15, 5))

#ç»˜åˆ¶æ­£æ€å›å½’æ¨¡å‹ä¸­ï¼Œé¢„æµ‹å˜é‡ä¸ºå‹åŠ›æƒ…å†µä¸‹çš„å…ˆéªŒé¢„æµ‹çº¿æ€§æ¨¡å‹
ax[0].plot(normal_prior.constant_data["stress"], 
           normal_prior.prior["mu"].stack(sample=("chain", "draw")), c="k", alpha=0.4)

#è®¾ç½®åæ ‡è½´æ ‡é¢˜
ax[0].set_title('Normal regression', fontsize=20) 
ax[0].set_xlabel('Stress', fontsize=16) 
ax[0].set_ylabel('Networksize', fontsize=16) 

#ç»˜åˆ¶ç»˜åˆ¶æ³Šæ¾å›å½’æ¨¡å‹ä¸­ï¼Œé¢„æµ‹å˜é‡ä¸ºå‹åŠ›æƒ…å†µä¸‹çš„å…ˆéªŒé¢„æµ‹çº¿æ€§æ¨¡å‹
ax[1].plot(poisson_prior.constant_data["stress"], 
           poisson_prior.prior["lam"].stack(sample=("chain", "draw")), c="k", alpha=0.4)

#è®¾ç½®åæ ‡è½´æ ‡é¢˜ã€yè½´åˆ»åº¦æ˜¾ç¤º
ax[1].set_title('Possion regression', fontsize=20) 
ax[1].set_xlabel('Stress', fontsize=16) 
ax[1].set_ylabel('Networksize', fontsize=16) 
ax[1].ticklabel_format(axis='y', style='plain')

sns.despine()
plt.show()
```

### æ¨¡å‹å®šä¹‰  

$$  
\begin{equation}  
\begin{array}{lcrl}  
\text{data:} & \hspace{.025in} & Y_i|\beta_0,\beta_1,\beta_2,r & \stackrel{ind}{\sim} \text{NegBin}\left(\mu_i, r \right) \;\; \text{ with } \;\; \mu_i= e^{\beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2}}\\  
\text{priors:} & & \beta_{0}  & \sim N\left(0, 0.5^2\right) \\  
 & & \beta_1  & \sim N\left(0, 0.1^2\right) \\  
 & & \beta_2  & \sim N\left(0, 0.5^2\right) \\  
& & r & \sim \text{Exp}(1)\\  
\end{array}  
\end{equation}  
$$

```python
with pm.Model(coords = {"obs_id": df.index}) as nb_model:

    beta_0 = pm.Normal("beta_0", mu=0, sigma=0.5)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=0.1)           #å®šä¹‰beta_1
    beta_2 = pm.Normal("beta_2", mu=0, sigma=0.5)           #å®šä¹‰beta_2
    alpha = pm.Exponential("alpha", 1)

    stress = pm.MutableData("stress",df.stress, dims="obs_id")                           #è‡ªå˜é‡1:å‹åŠ›æ°´å¹³
    romantic = pm.MutableData("romantic",df.romantic, dims="obs_id")                     #è‡ªå˜é‡2:æ€§åˆ«
    
    lam = pm.Deterministic("lam", 
                           pm.math.exp(beta_0 + beta_1*stress + beta_2*romantic), 
                           dims="obs_id")    #å®šä¹‰muï¼Œè‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ
    
    likelihood = pm.NegativeBinomial("y_est", mu=lam, alpha=alpha, observed=df.networksize, dims="obs_id")
```

## æ€»ç»“  

æˆ‘ä»¬ç°åœ¨å·²ç»å­¦ä¹ äº†æ­£æ€ã€æ³Šæ¾å’Œè´ŸäºŒé¡¹åˆ†å¸ƒæ•°æ®ç»“æ„çš„è´å¶æ–¯å›å½’æ¨¡å‹ï¼Œç„¶è€Œè¿˜å­˜åœ¨å…¶ä»–å½¢å¼çš„å›å½’æ¨¡å‹ï¼Œæ¯”å¦‚ä¼½é©¬ã€é€†æ­£æ€ã€å¯¹æ•°æ­£æ€ç­‰ã€‚æ‰€æœ‰è¿™äº›é€‰é¡¹éƒ½å±äºå¹¿ä¹‰çº¿æ€§æ¨¡å‹ï¼ˆ**generalized linear modelï¼ŒGLM**ï¼‰çš„ä¸€ä¸ªå¤§ç±»ã€‚  
* é¦–å…ˆï¼Œå¿…é¡»æ³¨æ„å› å˜é‡ Y çš„ç»“æ„ã€‚Y æ˜¯ç¦»æ•£çš„è¿˜æ˜¯è¿ç»­çš„ï¼Ÿæ˜¯å¯¹ç§°çš„è¿˜æ˜¯å€¾æ–œçš„ï¼ŸY çš„å–å€¼èŒƒå›´æ˜¯ä»€ä¹ˆï¼Ÿè¿™äº›é—®é¢˜å¯ä»¥å¸®åŠ©æˆ‘ä»¬ç¡®å®šåˆé€‚çš„æ•°æ®ç»“æ„ã€‚  
* å…¶æ¬¡ï¼Œå®šä¹‰æ•°æ®çš„æœŸæœ› E(Y âˆ£ ...)å’Œé“¾æ¥å‡½æ•° g(Â·)ã€‚  
	* å¯¹äºæ‰€æœ‰ GLMï¼ŒE(Y âˆ£ ...) å¯¹é¢„æµ‹å› å­çº¿æ€§ç»„åˆ (X1, X2, ..., Xp) çš„ä¾èµ–å…³ç³»è¡¨ç¤ºä¸º $g(E(Y \mid \ldots))=\beta_{0}+\beta_{1} X_{1}+\beta_{2} X_{2}+\cdots+\beta_{p} X_{p}$ã€‚å…¶ä¸­é“¾æ¥å‡½æ•° g(â‹…) å–å†³äºæ•°æ®ç»“æ„ã€‚  


| æ¨¡å‹     | æ­£æ€æ¨¡å‹      | æ³Šæ¾æ¨¡å‹      | è´ŸäºŒé¡¹æ¨¡å‹    |   
| -------- | ----------------- | ---------------- | ------------------ |  
| æ•°æ®æ¨¡å‹ | $Y_i\| \beta_0, \beta_1, \cdots, \beta_p, \sigma \sim N(\mu_i, \sigma^2)$ | $Y_i\| \beta_0, \beta_1, \cdots, \beta_p \sim Pois(\lambda_i)$   |   $Y_i\| \beta_0, \beta_1, \cdots, \beta_p, \sigma \sim NegBin(\mu_i, r)$    |   
| è¿æ¥å‡½æ•° | $g(\mu_i) = \mu_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_p X_{ip}$     | $g(\lambda_i) := \log(\lambda_i) = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_p X_{ip}$ |  $g(\mu_i) := \log(\mu_i) = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_p X_{ip}$  |   


æˆ‘ä»¬ä¸éœ€è¦å¯¹æ¯ä¸€ç§ GLM è¿›è¡Œéå¸¸ä»”ç»†çš„ç ”ç©¶ã€‚Bambi å·¥å…·æä¾›äº†å„ç§åˆ†å¸ƒå½¢æ„å»ºå¸®åŠ©ç”¨æˆ·æ„å»º GLMï¼Œä»¥æ­¤é€‚ç”¨äºä¸åŒçš„æ•°æ®åˆ†å¸ƒã€‚  

åœ¨åç»­çš„è¯¾ç¨‹ä¸­ï¼Œæˆ‘ä»¬å°†å¼€å§‹ä½¿ç”¨ Bambi æ¥æ„å»ºä¸åŒçš„è´å¶æ–¯æ¨¡å‹ã€‚

#### æ¨¡å‹å®šä¹‰

```python
# æˆ‘ä»¬éœ€è¦å¯¹å…ˆéªŒè¿›è¡Œå®šä¹‰ï¼Œåœ¨å®šä¹‰æ—¶éœ€è¦æ³¨æ„å‚æ•°çš„å‘½åï¼Œå¯ä»¥å…ˆåªä¼ å…¥è¡¨è¾¾å¼
# æŸ¥çœ‹åœ¨bambiä¸­å„å‚æ•°å¯¹åº”çš„å‘½åï¼Œæ¥ä¸‹æ¥è¿›è¡Œä¿®æ”¹
import bambi as bmb

bambi_default = bmb.Model("networksize ~ stress + romantic", 
                     df,
                     family="negativebinomial")
bambi_default 
```

```python
#å®šä¹‰å…ˆéªŒå¹¶ä¼ å…¥æ¨¡å‹ä¸­
intercept_prior = bmb.Prior("Normal", mu=0, sigma=0.5)     #å¯¹åº”beta_0
stress_prior = bmb.Prior("Normal", mu=0, sigma=0.1)        #å¯¹åº”beta_1
romantic_prior = bmb.Prior("Normal", mu=0, sigma=0.5)      #å¯¹åº”beta_2
alpha_prior = bmb.Prior("Exponential", lam = 1)            #å¯¹åº”alpha

# å°†ä¸‰ä¸ªå‚æ•°çš„å…ˆéªŒå®šä¹‰åœ¨å­—å…¸priorä¸­ï¼Œæ³¨æ„å‚æ•°çš„å‘½å
priors = {"Intercept": intercept_prior, 
          "stress": stress_prior,
          "romantic": romantic_prior,
          "alpha": alpha_prior}

#å®šä¹‰å…³ç³»å¼ï¼Œä¼ å…¥æ•°æ®
bambi_nb = bmb.Model("networksize ~ stress + romantic", 
                     df,
                     priors=priors, 
                     family="negativebinomial")

#æ€»ç»“å¯¹æ¨¡å‹çš„è®¾ç½®
bambi_nb
```

```python
bambi_nb_trace = bambi_nb.fit(draws=5000,                   # ä½¿ç”¨mcmcæ–¹æ³•è¿›è¡Œé‡‡æ ·ï¼Œdrawsä¸ºé‡‡æ ·æ¬¡æ•°
                              tune=1000,                    # tuneä¸ºè°ƒæ•´é‡‡æ ·ç­–ç•¥çš„æ¬¡æ•°ï¼Œå¯ä»¥å†³å®šè¿™äº›ç»“æœæ˜¯å¦è¦è¢«ä¿ç•™
                              chains=4,
                              random_seed=84735)
```
