```python
import seaborn as sns
import pandas as pd
```

```python
# å¯¼å…¥seabornåº“ä¸­å¸¦æœ‰çš„æ•°æ®é›†
penguins = pd.read_csv("/home/mw/input/37953795/penguins.csv")

#æŸ¥çœ‹å‰äº”è¡Œæ•°æ®
penguins.head()
```

## ç®±å›¾

```python
# æŒ‡å®šxè½´ä¸yè½´ä»£è¡¨çš„å˜é‡
sns.boxplot(data=penguins, 
            x="species", 
            y="flipper_length_mm")

# å»æ‰å³è¾¹å’Œä¸Šé¢çš„è¾¹æ¡†ï¼Œä½¿åæ ‡è½´æ›´ç¬¦åˆAPAè¦æ±‚
sns.despine()
```

## æ¡å½¢å›¾

```python
# æŒ‡å®šxè½´ä¸yè½´ä»£è¡¨çš„å˜é‡ï¼Œä½¿ç”¨hueæ¥æŒ‡å®šç¬¬äºŒä¸ªåˆ†ç±»æ ‡å‡†ï¼šæ€§åˆ«
sns.barplot(data = penguins, 
            x = "species", 
            y = "flipper_length_mm", 
            hue="sex",
            errorbar="sd")
sns.despine()
```

## æ•£ç‚¹å›¾

```python
# æŒ‡å®šxè½´ä¸yè½´ä»£è¡¨çš„å˜é‡
sns.scatterplot(data = penguins, 
                x = "bill_length_mm", 
                y = "bill_depth_mm")
sns.despine()
```

```python
# å¢åŠ æ‹Ÿåˆçš„å›å½’ç›´çº¿
sns.regplot(data = penguins, 
                x = "bill_length_mm", 
                y = "bill_depth_mm")
sns.despine()
```
