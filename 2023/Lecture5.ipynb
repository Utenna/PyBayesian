**è®¡ç®—Poissonçš„ä¼¼ç„¶å‡½æ•°**  


ğŸ¤”å¦‚æœæˆ‘ä»¬ä¸€å…±è·å¾—äº†å››å¤©çš„æ•°æ®ï¼Œæ­¤æ—¶çš„ä¼¼ç„¶å‡½æ•°æ€ä¹ˆå†™ï¼Ÿ  

æˆ‘ä»¬å°†å››å¤©å†…ä¸­å›½è·å¾—çš„é‡‘ç‰Œæ•°é‡$(Y_1, Y_2, \ldots, Y_n)$ç»„åˆä¸ºä¸€ä¸ªå‘é‡$\vec{y} = (y_1,y_2,\ldots,y_n)$  

åœ¨ç»™å®šçš„$\lambda$ä¸‹ï¼Œ$\vec{y}$å‘ç”Ÿçš„å¯èƒ½æ€§å¯ä»¥è¡¨ç¤ºä¸ºï¼š  

$$  
f(\vec{y} | \lambda) = \prod_{i=1}^n f(y_i | \lambda) = f(y_1 | \lambda) \cdot f(y_2 | \lambda) \cdot \cdots \cdot f(y_n | \lambda)  
$$  

> æ³¨æ„ï¼šç”±äºæ¯å¤©è·å¾—é‡‘ç‰Œçš„æ•°é‡æ˜¯äº’ä¸å½±å“çš„ï¼Œæ‰€ä»¥æˆ‘ä»¬å¯ä»¥ç›´æ¥ç›¸ä¹˜ï¼š$P(A \cap B) = P(A)P(B)$  


æˆ‘ä»¬å¯ä»¥æŠŠå…¬å¼ä»£å…¥å¹¶è¿›è¡Œæ•´ç†ï¼Œæœ€åå¯å¾—åˆ°  

$$  
\begin{split}  
f(\vec{y} | \lambda) =\frac{\lambda^{\sum y_i}e^{-n\lambda}}{\prod_{i=1}^n y_i!} \\  
\end{split}  
$$  

å…¶ä¸­nä¸ºæ•°æ®çš„æ•°é‡(åœ¨è¿™ä¸ªä¾‹å­ä¸­æ˜¯æ”¶é›†å¤šå°‘å¤©çš„æ•°æ®)ï¼Œå½“nå¾ˆå¤§æ—¶ï¼Œ$\prod y_i!$çš„è®¡ç®—å°†å˜å¾—éº»çƒ¦ï¼ŒåŒæ ·çš„æˆ‘ä»¬å¯ä»¥æš‚æ—¶å¿½ç•¥åˆ†æ¯  

$$  
L(\lambda | \vec{y}) = \frac{\lambda^{\sum y_i}e^{-n\lambda}}{\prod_{i=1}^n y_i!} \propto \lambda^{\sum y_i}e^{-n\lambda} \;\; \text{ for } \lambda > 0  
$$

**å…·ä½“è®¡ç®—**  

$$  
\begin{split}  
f(\vec{y} | \lambda)  
& = \frac{\lambda^{y_1}e^{-\lambda}}{y_1!} \cdot \frac{\lambda^{y_2}e^{-\lambda}}{y_2!} \cdots \frac{\lambda^{y_n}e^{-\lambda}}{y_n!} \\  
& = \frac{\left[\lambda^{y_1}\lambda^{y_2} \cdots \lambda^{y_n}\right] \left[e^{-\lambda}e^{-\lambda} \cdots e^{-\lambda}\right]}{y_1! y_2! \cdots y_n!} \\  
& =\frac{\lambda^{\sum y_i}e^{-n\lambda}}{\prod_{i=1}^n y_i!} \\  
\end{split}  
$$

**ä»£å…¥å…·ä½“æ•°æ®**  

å‡è®¾å››å¤©å†…ä¸­å›½è·å¾—é‡‘ç‰Œçš„æ•°é‡åˆ†åˆ«æ˜¯20ï¼Œ13ï¼Œ13ï¼Œ8ï¼Œæˆ‘ä»¬å¯ä»¥å°†å…¶ç»„åˆæˆä¸€ä¸ªå‘é‡$\vec{y} = (y_1,y_2,y_3,y_4) = (20ï¼Œ13ï¼Œ13ï¼Œ8)$  

ä»£å…¥å…¬å¼ï¼š  
$$  
L(\lambda | \vec{y}) = \frac{\lambda^{\sum y_i}e^{-n\lambda}}{\prod_{i=1}^n y_i!} \;\;\;\;\;\;\;\;\;\sum_{i=1}^4y_i = 20 + 13 + 13 + 8 = 54  
$$  

å¯å¾—ï¼š  
$$  
L(\lambda | \vec{y}) = \frac{\lambda^{54}e^{-4\lambda}}{20!\times13!\times13!\times8!} \propto \lambda^{54}e^{-4\lambda} \;\;\;\; \text{ for } \lambda > 0  
$$  

æ­¤æ—¶ï¼Œè¾“å…¥ä¸åŒçš„å‚æ•°$\lambda$ï¼Œå°±å¯ä»¥è®¡ç®—å¾—åˆ°å¯¹åº”å‚æ•°çš„ä¼¼ç„¶å€¼ã€‚

é¦–å…ˆï¼Œæˆ‘ä»¬å°è¯•ç»˜åˆ¶æ•°æ®çš„é¢‘ç‡åˆ†å¸ƒå›¾ï¼Œä»¥ç¡®å®šæ•°æ®çš„å½¢æ€ã€‚  

å…¶æ¬¡ï¼Œæˆ‘ä»¬å‡è®¾ä¸­å›½æ¯å¤©å¹³å‡è·å¾—12æšé‡‘ç‰Œ($\lambda=12$)ï¼Œå¹¶å°è¯•è®¡ç®—å¯¹åº”æ•°æ®çš„ä¼¼ç„¶å€¼ã€‚

```python
# ç»˜åˆ¶æ•°æ®çš„åˆ†å¸ƒ

#------------------------------------------------
#                ä½ å¯ä»¥æ›´æ”¹ä»¥ä¸‹æ•°æ®ï¼Œä½“ä¼šä¸åŒæ•°æ®å½¢æˆçš„åˆ†å¸ƒï¼Œä¾‹å¦‚ observed_data = [3,3,2,2,2,1,...]
#------------------------------------------------
observed_data = [20,13,13,8]   # å‡è®¾è¿ç»­å››å¤©ä¸­å›½è·å¾—é‡‘ç‰Œçš„æ•°é‡

plt.hist(observed_data)
plt.title('Histogram of Observations')
plt.xlabel('Number of Gold Medal')
plt.ylabel('Frequency')
plt.show()
```

**Poissonä¼¼ç„¶å‡½æ•°å›¾ç¤º**  

å¦‚ä½•ç”»å‡ºPoissonçš„ä¼¼ç„¶å‡½æ•°å›¾ï¼Ÿ  

- æˆ‘ä»¬éœ€è¦æ¨ªè½´ä¸ºä¸åŒçš„$\lambda$ï¼Œçºµè½´ä¸ºåœ¨ä¸åŒçš„$\lambda$ä¸‹ï¼Œ$\vec{y}$å‘ç”Ÿçš„ä¼¼ç„¶$L(\lambda | \vec{y})$ï¼š

```python
# å®šä¹‰ä¼¼ç„¶å‡½æ•°ï¼Œç”±äºè¯¥ä¼¼ç„¶å‡½æ•°ä¸èƒ½é€šè¿‡è°ƒç”¨å·²æœ‰çš„å‡½æ•°è·å¾—ï¼Œå› æ­¤æˆ‘ä»¬æ ¹æ®å¾—åˆ°çš„å…¬å¼å®šä¹‰ä¸€ä¸ª

def poisson_likelihood(y):
    lambdas = np.arange(0, 25, 0.1)
    
    # Calculate the Poisson likelihood for each lambda
    likelihood_values = np.exp(-len(y) * lambdas) * np.power(lambdas, np.sum(y)) / np.prod([np.math.factorial(val) for val in y])

    return lambdas, likelihood_values

#ä¼ å…¥éœ€è¦çš„å‚æ•°vec y
observed_data = [20,13,13,8]
lambdas, likelihood_values = poisson_likelihood(observed_data)
likelihood_values /= np.sum(likelihood_values)

plt.plot(lambdas, likelihood_values, color='black', lw=2)
plt.xlabel('$\lambda$')
plt.ylabel('$L(\lambda | Y=y)$')
sns.despine()
```

###  Potential priors

**ğŸ¤”æ€è€ƒï¼š**  

åœ¨ä¹‹å‰çš„ä¾‹å­ä¸­ï¼Œæˆ‘ä»¬ä¸»è¦å…³æ³¨æ³Šæ¾åˆ†å¸ƒç›¸å…³çš„ä¼¼ç„¶å‡½æ•°ï¼Œè€Œå¿½ç•¥äº†å‚æ•°$\lambda$çš„å…ˆéªŒåˆ†å¸ƒã€‚  
$$  
L(\lambda | \vec{y}) = \frac{\lambda^{\sum y_i}e^{-n\lambda}}{\prod_{i=1}^n y_i!} \propto \lambda^{11}e^{-4\lambda}  
$$

```python
# @title interactive gamma
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import gamma
from ipywidgets import interact
import seaborn as sns


x = np.linspace(0, 20, 1000)
def plot_beta(a, b):
    y = gamma.pdf(x, a=a, scale=1/b)
    plt.plot(x, y, color='black', lw=2)
    plt.title(f"Gamma({a}, {b})")
    plt.xlabel('$\lambda$')
    plt.ylabel('$f(\lambda)$')
    plt.gca().yaxis.set_major_formatter('{x:1.1f}')
    plt.xlim(0, 8)
    plt.xticks(np.arange(0, 9, 2))
    plt.yticks(np.arange(0, 2.5, 0.5))
    sns.despine()
```

### Gamma-Poisson conjugacy

ä¼¼ç„¶ï¼š& Y_i | \lambda \stackrel{ind}{\sim} \text{Pois}(\lambda) \\  
åéªŒï¼š& \lambda|\vec{y} \; \sim \; \text{Gamma}\left(s + \sum y_i, \; r + n\right)  
\end{align*}  
$$

### The Normal data model  

åŒæ ·ï¼Œæˆ‘ä»¬çš„è´å¶æ–¯åˆ†æä¹Ÿæ˜¯ä»å¯¹å¹³å‡èº«é«˜(Î¼)çš„å…ˆéªŒå¼€å§‹ã€‚  
ç„¶è€Œï¼Œè¦ä¸ºÎ¼æŒ‡å®šä¸€ä¸ªé€‚å½“çš„å…ˆéªŒæ¨¡å‹ç»“æ„ï¼Œéœ€è¦è€ƒè™‘æ•°æ®(Yi)çš„ç‰¹å¾ã€‚  

ç”±äºèº«é«˜ Yi æ˜¯è¿ç»­æ•°æ®ï¼Œå› æ­¤æœ‰å¾ˆå¤šæ½œåœ¨å¯èƒ½çš„æ¨¡å‹ï¼šè´å¡”æ¨¡å‹ã€æŒ‡æ•°æ¨¡å‹ã€ä¼½é©¬æ¨¡å‹ã€æ­£æ€åˆ†å¸ƒæ¨¡å‹ã€F æ¨¡å‹ç­‰ã€‚  
- ä»è¿™ä¸ªåˆ—è¡¨ä¸­ï¼Œæˆ‘ä»¬å¯ä»¥ç«‹å³æ’é™¤è´å¡”æ¨¡å‹ã€‚å› ä¸ºå®ƒå‡è®¾ Yi âˆˆ [0ï¼Œ1]ï¼Œè€Œèº«é«˜å¾€å¾€åœ¨ 165cm å·¦å³ã€‚  
- åœ¨å‰©ä¸‹çš„é€‰é¡¹ä¸­ï¼Œæ­£æ€æ¨¡å‹æ˜¯ç›¸å½“åˆç†çš„ã€‚è¿™ä¹Ÿæ˜¯æˆ‘ä»¬ç›´è§‰ä¸Šçš„ç¬¬ä¸€ååº”ï¼Œæ¯•ç«Ÿèº«é«˜é€šå¸¸å›´ç»•æŸä¸ªæ€»ä½“å¹³å‡å€¼ï¼ˆÎ¼ï¼‰å‘ˆç°ä¸ºå¯¹ç§°æˆ–æ­£æ€åˆ†å¸ƒã€‚

**æ­£æ€æ¨¡å‹ä»‹ç»**  

å‡è®¾ Y æ˜¯ä¸€ä¸ªè¿ç»­éšæœºå˜é‡ï¼Œå¯ä»¥å– -âˆ å’Œ âˆ ä¹‹é—´çš„ä»»æ„å€¼ï¼Œå³ Yâˆˆ (-âˆ, âˆ)ã€‚  

é‚£ä¹ˆï¼ŒY çš„å˜å¼‚æ€§å¯ä»¥å¾ˆå¥½åœ°ç”¨å‡å€¼å‚æ•° Î¼âˆˆ (-âˆ, âˆ) å’Œæ ‡å‡†å·®å‚æ•° Ïƒ > 0 çš„æ­£æ€æ¨¡å‹æ¥è¡¨ç¤ºï¼š  

$$  
Y \sim N(\mu, \sigma^2).  
$$  

æ­£æ€æ¨¡å‹çš„è¿ç»­ pdf å½¢å¼:  

$$  
\begin{equation}  
f(y) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\bigg[{-\frac{(y-\mu)^2}{2\sigma^2}}\bigg] \;\; \text{ for } y \in (-\infty,\infty)  
\tag{5.11}  
\end{equation}  
$$  

æ­£å¤ªæ¨¡å‹çš„åŸºæœ¬ç‰¹å¾ï¼š  

$$  
\begin{split}  
E(Y) & = \text{ Mode}(Y) = \mu \\  
\text{Var}(Y) & = \sigma^2 \\  
\text{SD}(Y) & = \sigma. \\  
\end{split}  
$$  

> æ­¤å¤–ï¼Œå¤§å®¶éƒ½çŸ¥é“çš„ï¼Œå¤§çº¦ 95% çš„ Y å€¼éƒ½åœ¨ Î¼ çš„ 2 ä¸ªæ ‡å‡†å·®èŒƒå›´å†…ã€‚

```python
import seaborn as sns
import matplotlib.pyplot as plt

# åˆ›å»ºæ•°æ®
import numpy as np
np.random.seed(0)
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=5, scale=2, size=1000)
data3 = np.random.normal(loc=10, scale=5, size=1000)

# ç»˜åˆ¶å›¾å½¢
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# ç»˜åˆ¶ç¬¬ä¸€ä¸ªå›¾
sns.histplot(data=data1, kde=True, ax=axes[0], stat = 'density')
axes[0].set_title('Mean=0, Std=1')
axes[0].set_ylim([0, 0.5])

# ç»˜åˆ¶ç¬¬äºŒä¸ªå›¾
sns.histplot(data=data2, kde=True, ax=axes[1], stat = 'density')
axes[1].set_title('Mean=5, Std=2')
axes[1].set_ylim([0, 0.5])

# ç»˜åˆ¶ç¬¬ä¸‰ä¸ªå›¾
sns.histplot(data=data3, kde=True, ax=axes[2], stat = 'density')
axes[2].set_title('Mean=10, Std=5')
axes[2].set_ylim([0, 0.5])

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
# ç§»é™¤å›¾çš„ä¸Šã€å³è¾¹æ¡†çº¿
sns.despine()
```

æ ¹æ®2020å¹´ä¹‹å‰çš„æ•°æ®ï¼Œæˆ‘ä»¬å¯ä»¥ç”¨ 165cm ä½œä¸ºæˆ‘ä»¬æ­£æ€å…ˆéªŒæ¨¡å‹çš„å‡å€¼ï¼Œå³ Âµ = 165ã€‚

$$  
\begin{align*}  
\mu & \sim N(\theta, \tau^2) \\

\tag{5.14}  
\end{align*}  
$$

$$  
\exp\bigg[{-\frac{(\mu - \blacksquare)^2}{2\blacksquare^2}}\bigg]  
$$

### Normal-Normal conjugacy  

ä¸ºäº†å¾—åˆ° Î¼ çš„åéªŒæ¨¡å‹ï¼Œæˆ‘ä»¬å¿…é¡»å°†å…ˆéªŒæ¨¡å‹å’Œæ•°æ®ä¿¡æ¯ç»“åˆèµ·æ¥ã€‚  

> éœ€è¦å¼ºè°ƒçš„æ˜¯ï¼Œæˆ‘ä»¬é€‰æ‹©çš„æ­£æ€å…ˆéªŒæ¨¡å‹å±äºæ­£æ€-æ­£æ€çš„å…±è½­æ—ï¼å› æ­¤ï¼ŒÎ¼ çš„åéªŒæ¨¡å‹ä¹Ÿå°†æ˜¯æ­£æ€çš„ï¼Œå¹¶æ ¹æ®å…ˆéªŒæ•°æ®å’Œè§‚æµ‹æ•°æ®æ›´æ–°å‚æ•°ã€‚  

ä»¥ä¸‹å…¬å¼å±•ç¤ºäº†Î¼çš„åéªŒæ¨¡å‹ï¼š  

$$  
\begin{align*}  
Y_i | \mu & \stackrel{ind}{\sim} N(\mu, \sigma^2) \\  
\mu & \sim N(\theta, \tau^2) \\  
\mu|\vec{y} \; & \sim \;  N\bigg(\theta\frac{\sigma^2}{n\tau^2+\sigma^2} + \bar{y}\frac{n\tau^2}{n\tau^2+\sigma^2}, \; \frac{\tau^2\sigma^2}{n\tau^2+\sigma^2}\bigg)  
\tag{5.15}  
\end{align*}  
$$  

é¦–å…ˆï¼ŒåéªŒå‡å€¼æ˜¯å…ˆéªŒå‡å€¼ E(Î¼) = Î¸ å’Œæ ·æœ¬å‡å€¼ y çš„åŠ æƒå¹³å‡å€¼ï¼›  

$$  
\text{posterior mean} \sim \theta\frac{\sigma^2}{n\tau^2+\sigma^2} + \bar{y}\frac{n\tau^2}{n\tau^2+\sigma^2}  
$$  

å…¶æ¬¡ï¼ŒåéªŒæ–¹å·®å—å…ˆéªŒå˜å¼‚æ€§ Ï„ å’Œæ•°æ®å˜å¼‚æ€§ Ïƒ çš„å½±å“ï¼š  

$$  
\text{posterior variance}~\frac{\tau^2\sigma^2}{n\tau^2+\sigma^2}  
$$  

ä¸¤è€…éƒ½å—åˆ°æ ·æœ¬é‡nçš„å½±å“ã€‚  
- é¦–å…ˆï¼Œéšç€nçš„å¢åŠ ï¼ŒåéªŒå‡å€¼å¯¹å…ˆéªŒå‡å€¼çš„æƒé‡å‡å°‘ï¼Œå¯¹æ ·æœ¬å‡å€¼yçš„æƒé‡å¢åŠ : $\frac{\sigma^2}{n\tau^2+\sigma^2} \to 0 \;\; \text{ and } \;\;\frac{n\tau^2}{n\tau^2+\sigma^2} \to 1 .$  
- æ­¤å¤–ï¼Œéšç€ n çš„å¢å¤§ï¼ŒåéªŒæ–¹å·®ä¼šå‡å° $\frac{\tau^2\sigma^2}{n\tau^2+\sigma^2} \to 0$  
- ä¹Ÿå°±æ˜¯è¯´ï¼Œæˆ‘ä»¬æŒæ¡çš„æ•°æ®è¶Šå¤šï¼Œæˆ‘ä»¬å¯¹ Î¼ çš„åéªŒç¡®å®šæ€§å°±è¶Šé«˜ï¼Œä¹Ÿå°±è¶Šä¸æ•°æ®çš„ç»“æœç›¸ä¼¼ã€‚

ç»“åˆä¹‹å‰çš„å…ˆéªŒæ¨¡å‹å’Œæ•°æ®æ¨¡å‹ï¼Œç°åœ¨æˆ‘ä»¬å¯ä»¥æ¥è®¡ç®—åéªŒå¹¶ç»˜åˆ¶å®ƒã€‚  

- å…ˆéªŒæ¨¡å‹å‡è®¾å…¨å›½æˆäººçš„å¹³å‡èº«é«˜ä¸º 165cmï¼Œæ ‡å‡†åå·®ä¸º 5cmã€‚  
- æ•°æ®æ¨¡å‹ä¸º 10000 åè¢«è¯•çš„èº«é«˜æ•°æ®ï¼Œå¹³å‡èº«é«˜ä¸º 168cmï¼Œæ ‡å‡†åå·®ä¸º 5cmã€‚

```python
# å®šä¹‰èº«é«˜èŒƒå›´
x = np.linspace(150, 200, 10000)

# å®šä¹‰å…ˆéªŒåˆ†å¸ƒ
prior_mean = 165
prior_std = 5
prior_y = st.norm.pdf(x, loc = prior_mean, scale = prior_std) / np.sum(st.norm.pdf(x, prior_mean, prior_std))

# ç”Ÿæˆä¼¼ç„¶
likelihood_mean = 168
likelihood_std = 5
likelihood_values = st.norm.pdf(x, loc = likelihood_mean, scale = likelihood_std) / np.sum(st.norm.pdf(x, likelihood_mean, likelihood_std))

# è®¡ç®—åéªŒåˆ†å¸ƒ
posterior_mean = (prior_mean * prior_std**2 + likelihood_mean * likelihood_std**2) / (prior_std**2 + likelihood_std**2)
posterior_std = np.sqrt((prior_std**2 * likelihood_std**2) / (prior_std**2 + likelihood_std**2))
posterior = st.norm.pdf(x, loc = posterior_mean, scale = posterior_std) / np.sum(st.norm.pdf(x, posterior_mean, posterior_std))


plt.plot(x, prior_y, color='#f0e442', label="prior")
plt.fill_between(x, prior_y, color="#f0e442", alpha=0.5)
plt.plot(x, likelihood_values, color='#0071b2', label="likelihood")
plt.fill_between(x, likelihood_values, color="#0071b2", alpha=0.5)
plt.plot(x, posterior, color='#009e74', label="posterior")
plt.fill_between(x, posterior, color="#009e74", alpha=0.5)

# è®¾ç½® x å’Œ y è½´æ ‡ç­¾
plt.xlabel('$\mu$ for height')
plt.ylabel('density')
plt.legend()

# ç§»é™¤å›¾çš„ä¸Šã€å³è¾¹æ¡†çº¿
sns.despine()
```

**æ‰©å±•é˜…è¯»ï¼šè¯æ˜ Normal-Normal conjugacy**  

ä¸ºå®Œæ•´èµ·è§ï¼Œæˆ‘ä»¬åœ¨æ­¤è¯æ˜æ­£æ€-æ­£æ€æ¨¡å‹äº§ç”ŸåéªŒæ¨¡å‹ã€‚å¦‚æœä½ ä¸å–œæ¬¢æ¨å¯¼ï¼Œä¹Ÿæ²¡å…³ç³»ã€‚è¯·éšæ„è·³åˆ°ä¸‹ä¸€èŠ‚ã€‚  

è®©æˆ‘ä»¬ç›´æ¥è¿›å…¥æ­£é¢˜ã€‚Î¼ çš„åéªŒ pdf ä¸æ­£æ€å…ˆéªŒ pdf å’Œä¼¼ç„¶å‡½æ•° çš„ä¹˜ç§¯æˆæ­£æ¯”ã€‚å¯¹äº Î¼âˆˆ (-âˆ, âˆ)ï¼š  
$$  
f(\mu|\vec{y}) \propto f(\mu)L(\mu|\vec{y}) \propto \exp\bigg[{\frac{-(\mu - \theta)^2}{2\tau^2}}\bigg] \cdot \exp\bigg[{-\frac{(\bar{y}-\mu)^2}{2\sigma^2/n}}\bigg]  .  
$$  

æ¥ä¸‹æ¥ï¼Œæˆ‘ä»¬å¯ä»¥å±•å¼€æŒ‡æ•°ä¸­çš„å¹³æ–¹ï¼ŒåŒ…æ‹¬ç¬¬ä¸€ä¸ªæŒ‡æ•°åˆ†å­ä¸­çš„$\theta^2$ å’Œç¬¬äºŒä¸ªæŒ‡æ•°åˆ†å­ä¸­çš„ $y^2$ï¼š  
$$  
\begin{split}  
f(\mu|\vec{y})  
& \propto  \exp\Bigg[{\frac{-\mu^2+2\mu\theta-\theta^2}{2\tau^2}}\Bigg]\exp\Bigg[{\frac{-\mu^2+2\mu\bar{y}-\bar{y}^2}{2\sigma^2/n}}\Bigg] \\  
& \propto  \exp\Bigg[{\frac{-\mu^2+2\mu\theta}{2\tau^2}}\Bigg]\exp\Bigg[{\frac{-\mu^2+2\mu\bar{y}}{2\sigma^2/n}}\Bigg]. \\  
\end{split}  
$$  

æˆ‘ä»¬ç»™å‡ºæŒ‡æ•°çš„å…±åŒåˆ†æ¯ï¼Œå¹¶å°†å®ƒä»¬åˆå¹¶æˆä¸€ä¸ªæŒ‡æ•°ï¼š  
$$  
\begin{split}  
f(\mu|\vec{y})  
& \propto  \exp\Bigg[{\frac{(-\mu^2+2\mu\theta)\sigma^2/n}{2\tau^2\sigma^2/n}}\Bigg]\exp\Bigg[{\frac{(-\mu^2+2\mu\bar{y})\tau^2}{2\tau^2\sigma^2/n}}\Bigg] \\  
& \propto  \exp\Bigg[{\frac{(-\mu^2+2\mu\theta)\sigma^2 +(-\mu^2+2\mu\bar{y})n\tau^2}{2\tau^2\sigma^2}}\Bigg]. \\  
\end{split}  
$$  

ç°åœ¨ï¼Œè®©æˆ‘ä»¬æŠŠ Î¼ é¡¹åˆå¹¶èµ·æ¥ï¼Œé‡æ–°æ’åˆ—ï¼Œè¿™æ · $\mu^2$ å°±æ˜¯ï¼š  
$$  
\begin{split}  
f(\mu|\vec{y})  
& \propto  \exp\Bigg[{\frac{-\mu^2(n\tau^2+\sigma^2)+2\mu(\theta\sigma^2+ \bar{y}n\tau^2) }{2\tau^2\sigma^2}}\Bigg] \\  
& \propto  \exp\Bigg[{\frac{-\mu^2+2\mu\left(\frac{\theta\sigma^2 + \bar{y}n\tau^2}{n\tau^2+\sigma^2}\right) }{2(\tau^2\sigma^2) /(n\tau^2+\sigma^2)}}\Bigg]. \\  
\end{split}  
$$  

ä»”ç»†è§‚å¯Ÿï¼Œä½ ä¼šå‘ç°æˆ‘ä»¬å¯ä»¥å¸¦å›ä¸€äº›ä¸ä¾èµ–äº Î¼ çš„å¸¸æ•°æ¥å®Œæˆåˆ†å­ä¸­çš„å¹³æ–¹ï¼š  
$$  
\begin{split}  
f(\mu|\vec{y})  
& \propto  \exp\Bigg[{\frac{-\bigg(\mu - \frac{\theta\sigma^2 + \bar{y}n\tau^2}{n\tau^2+\sigma^2}\bigg)^2 }{2(\tau^2\sigma^2) /(n\tau^2+\sigma^2)}}\Bigg]. \\  
\end{split}  
$$  

è¿™çœ‹èµ·æ¥ä»ç„¶å¾ˆæ··ä¹±ï¼Œä½†æ˜¯ä¸€æ—¦æˆ‘ä»¬å®Œæˆäº†æ­£æ–¹å½¢ï¼Œæˆ‘ä»¬å®é™…ä¸Šå°±å¾—åˆ°äº† Î¼ çš„æ­£æ€ pdf å†…æ ¸ï¼Œå³ expéƒ¨åˆ†ã€‚é€šè¿‡ç¡®å®šç¼ºå¤±çš„éƒ¨åˆ† âˆï¼Œæˆ‘ä»¬å¯ä»¥å¾—å‡ºä»¥ä¸‹ç»“è®º  
$$  
\mu|\vec{y} \;  \sim  \; N\left(\frac{\theta\sigma^2+ \bar{y}n\tau^2}{n\tau^2+\sigma^2}, \;{\frac{\tau^2\sigma^2}{n\tau^2+\sigma^2}} \right)  
$$  

æˆ‘ä»¬å¯ä»¥å°†åéªŒå‡å€¼é‡ç»„ä¸ºå…ˆéªŒå‡å€¼ Î¼ å’Œæ•°æ®å‡å€¼ y çš„åŠ æƒå¹³å‡å€¼ï¼š  
$$  
\frac{\theta\sigma^2+ \bar{y}n\tau^2}{n\tau^2+\sigma^2} = \theta\frac{\sigma^2}{n\tau^2+\sigma^2} + \bar{y}\frac{n\tau^2}{n\tau^2+\sigma^2} .  
$$
