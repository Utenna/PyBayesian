### Median absolute error (MAE)

ç»å¯¹è¯¯å·®çš„ä¸­ä½æ•°ï¼Œmedian absolute error (MAE)è¡¡é‡äº†é¢„æµ‹è§‚æµ‹å€¼$Y_i$ä¸åéªŒé¢„æµ‹å‡å€¼ä¹‹é—´$Y_i'$çš„å·®å¼‚ã€‚

$$
\text{MAE} = median\left(|Y_i - \hat{Y}_i|\right)
$$

- å‡è®¾ $Y_1, Y_2â€¦, Y_n$è¡¨ç¤ºnä¸ªè§‚å¯Ÿç»“æœã€‚

- æ¯ä¸ª $Y_i$ éƒ½æœ‰å¯¹åº”çš„åéªŒé¢„æµ‹å€¼ï¼Œå…¶å‡å€¼ä¸º $Y_i'$

- é¦–å…ˆï¼Œä»¥å‹åŠ› $X_i=45$ ä¸‹çš„åéªŒé¢„æµ‹å€¼ä¸ºä¾‹ã€‚

- å†æ¬¡ï¼Œè®¡ç®—æ‰€æœ‰è‡ªå˜é‡ä¸‹ï¼ŒåéªŒé¢„æµ‹å‡å€¼å’Œè§‚æµ‹å€¼ä¹‹é—´çš„è¯¯å·®ã€‚

1. è¯»å–æ•°æ® Data_Sum_HPP_Multi_Site_Shareï¼Œé€‰æ‹©"Tsinghua"ç«™ç‚¹çš„æ•°æ®ä½œä¸ºç¤ºä¾‹ã€‚

2. ä½¿ç”¨ pymc å¯¹å‹åŠ›å˜é‡`stress` å’Œè‡ªæˆ‘æ§åˆ¶å˜é‡ `scontrol` å»ºç«‹å›å½’æ¨¡å‹ã€‚

3. åœ¨æ¨¡å‹æ‹Ÿåˆå(ç•¥è¿‡æ¨¡å‹è¯Šæ–­ç¯å¢ƒ)ï¼Œç”Ÿæˆæ¨¡å‹çš„åéªŒé¢„æµ‹ã€‚

4. æœ€åï¼Œè®¡ç®—å‹åŠ› $X_i=45$æ¡ä»¶ä¸‹ï¼ŒåéªŒé¢„æµ‹å‡å€¼å’Œè§‚æµ‹å€¼ä¹‹é—´çš„è¯¯å·®ã€‚

*(æ³¨ï¼šç”±äºå‹åŠ›åˆ†æ•°ä¸º45æ—¶ï¼Œå­˜åœ¨å¤šä¸ªè‡ªæˆ‘æ§åˆ¶åˆ†æ•°ä¸ä¹‹å¯¹åº”ï¼Œæˆ‘ä»¬ä»ä¸­éšæœºæŠ½å–ä¸€ä¸ªä½œä¸ºæ¼”ç¤º)*

```python
# å¯¼å…¥ pymc æ¨¡å‹åŒ…ï¼Œå’Œ arviz ç­‰åˆ†æå·¥å…· 
import pymc as pm
import arviz as az
import seaborn as sns
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd

# å¿½ç•¥ä¸å¿…è¦çš„è­¦å‘Š
import warnings
warnings.filterwarnings("ignore")
```

```python
# é€šè¿‡ pd.read_csv åŠ è½½æ•°æ® Data_Sum_HPP_Multi_Site_Share.csv
df_raw = pd.read_csv('Data_Sum_HPP_Multi_Site_Share.csv')

# ç­›é€‰ç«™ç‚¹ä¸º"Tsinghua"çš„æ•°æ®
df = df_raw[df_raw["Site"] == "Tsinghua"]

# é€‰æ‹©éœ€è¦çš„ä¸¤åˆ—
df = df[["stress","scontrol"]]

#è®¾ç½®ç´¢å¼•
df["index"] = range(len(df))
df = df.set_index("index")
```

```python
# ä»å‹åŠ›å€¼ä¸º45å¯¹åº”çš„è‡ªæˆ‘æ§åˆ¶åˆ†æ•°ä¸­éšæœºæŠ½å–ä¸€ä¸ªï¼Œå¹¶æ ‡è®°å…¶ç´¢å¼•å·
# æˆ‘ä»¬å°†æ ¹æ®è¯¥ç´¢å¼•å·æ¥æå–å¯¹åº”çš„åéªŒé¢„æµ‹å€¼
print(df[df.stress == 45].sample(n=1, random_state=84735))
```

```python
# å¯¹å‹åŠ›å˜é‡`stress` å’Œè‡ªæˆ‘æ§åˆ¶å˜é‡ `scontrol` å»ºç«‹å›å½’æ¨¡å‹
with pm.Model() as linear_model:

    beta_0 = pm.Normal("beta_0", mu=70, sigma=4)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=0.8)         #å®šä¹‰beta_1
    sigma = pm.Exponential("sigma", 0.6)                  #å®šä¹‰sigma

    #xæ˜¯è‡ªå˜é‡å‹åŠ›æ°´å¹³
    x = pm.MutableData("x",df.stress,dims="obs_id")       
    #å®šä¹‰muï¼Œè®²è‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ              
    mu = pm.Deterministic("mu", beta_0 + beta_1 * x, dims="obs_id")    
    #å®šä¹‰ä¼¼ç„¶ï¼šé¢„æµ‹å€¼yç¬¦åˆN(mu, sigma)åˆ†å¸ƒ, observed ä¸ºä¼ å…¥å®é™…æ•°æ®y è‡ªæˆ‘æ§åˆ¶æ°´å¹³
    likelihood = pm.Normal("y_est", mu=mu, sigma=sigma, observed=df.scontrol, dims="obs_id")   
```

```python
# æ¨¡å‹æ‹Ÿåˆ 
#===========================
#     æ³¨æ„ï¼ï¼ï¼ä»¥ä¸‹ä»£ç å¯èƒ½éœ€è¦è¿è¡Œ1-2åˆ†é’Ÿå·¦å³
#===========================
with linear_model:
    trace = pm.sample(draws=5000,                   # ä½¿ç”¨mcmcæ–¹æ³•è¿›è¡Œé‡‡æ ·ï¼Œdrawsä¸ºé‡‡æ ·æ¬¡æ•°
                      tune=1000,                    # tuneä¸ºè°ƒæ•´é‡‡æ ·ç­–ç•¥çš„æ¬¡æ•°ï¼Œå¯ä»¥å†³å®šè¿™äº›ç»“æœæ˜¯å¦è¦è¢«ä¿ç•™
                      chains=4,                     # é“¾æ•°
                      discard_tuned_samples= True,  # tuneçš„ç»“æœå°†åœ¨é‡‡æ ·ç»“æŸåè¢«ä¸¢å¼ƒ
                      random_seed=84735)
```

```python
with linear_model:
    ppc = pm.sample_posterior_predictive(trace,
                                         random_seed=84735)
```

    - `constant_data`å¯¹åº”åŸå§‹æ•°æ®é›†ä¸­çš„è‡ªå˜é‡ï¼Œå‹åŠ›åˆ†æ•°(df.scontrol)

**è®¡ç®—æ‰€æœ‰è¯¯å·®çš„ä¸­ä½æ•°(MAE)**

$$
{\mathsf{MAE}}={\mathsf{median}}|Y_{i}-Y_{i}^{\prime}|.

```python
from statistics import median

# è®¡ç®—é¢„æµ‹è¯¯å·®
pre_y_mean["pre_error"] = pre_y_mean["scontrol_original"] - pre_y_mean["scontrol_ppc_mean"]

# æœ€åï¼Œè®¡ç®—é¢„æµ‹è¯¯å·®çš„ä¸­ä½æ•°
MAE = median(pre_y_mean.pre_error)
print(f"è®¡ç®—å¾—åˆ°çš„MAE: {MAE:.2f}")
```

## ç»ƒä¹ 
* æˆ‘ä»¬ä½¿ç”¨ä¸€ä¸ªæ–°ä¾‹å­æ¥å·©å›ºå¯¹**MAE**ä¸**åéªŒé¢„æµ‹åŒºé—´**çš„è®¡ç®—

* ç±»åˆ«å˜é‡åŒæ ·ä¹Ÿå¯ä»¥ä½œä¸ºè‡ªå˜é‡

* è¿™é‡Œï¼Œæˆ‘ä»¬è€ƒè™‘ä¸€ä¸ªç®€å•çš„äºŒåˆ†å˜é‡*æŠ½çƒŸ*ï¼Œå°†å…¶ä½œä¸ºè‡ªå˜é‡ï¼Œæ¥é¢„æµ‹*è‡ªæˆ‘æ§åˆ¶åˆ†æ•°*

```python
# é€šè¿‡ pd.read_csv åŠ è½½æ•°æ® Data_Sum_HPP_Multi_Site_Share.csv
df_raw = pd.read_csv('Data_Sum_HPP_Multi_Site_Share.csv')

# ç­›é€‰ç«™ç‚¹ä¸º"Tsinghua"çš„æ•°æ®
df_METU = df_raw[df_raw["Site"] == "METU"]

# é€‰æ‹©éœ€è¦çš„ä¸¤åˆ—
df_METU = df_METU[["stress","scontrol","smoke"]]

#1 è¡¨ç¤ºæŠ½çƒŸï¼Œ2è¡¨ç¤ºä¸æŠ½çƒŸ---é‡æ–°ç¼–ç ---0 è¡¨ç¤ºä¸æŠ½çƒŸï¼Œ1è¡¨ç¤ºæŠ½çƒŸ
df_METU["smoke"] =  np.where(df_METU['smoke'] == 2, 0, 1)
df_METU["smoke_recode"] =  np.where(df_METU['smoke'] == 1, "yes", "no")

#è®¾ç½®ç´¢å¼•
df_METU["index"] = range(len(df_METU))
df_METU = df_METU.set_index("index")
```

* åœ¨è¿™ä¸ªä¾‹å­ä¸­ï¼Œæˆ‘ä»¬å°†æ˜¯å¦æŠ½çƒŸä½œä¸ºè‡ªå˜é‡ï¼Œå°†è‡ªæˆ‘æ§åˆ¶åˆ†æ•°ä½œä¸ºå› å˜é‡

    * åœ¨è‡ªå˜é‡ä¸­ï¼Œ`0` è¡¨ç¤ºä¸æŠ½çƒŸï¼Œ`1`è¡¨ç¤ºæŠ½çƒŸ

$$
X_{i2} = \begin{cases}
1 & \text{Yes} \\
0 & \text{No} \\
\end{cases}
$$

* ç®€å•ç”»å‡ºä¸¤ç§æƒ…å†µä¸‹ï¼Œè‡ªæˆ‘æ§åˆ¶åˆ†æ•°åˆ†å¸ƒçš„å¯†åº¦å›¾


![Image Name](https://cdn.kesci.com/upload/s40nk3kagu.png?imageView2/0/w/960/h/960)

* åŒæ ·åœ°ï¼Œæˆ‘ä»¬å¯ä»¥ä¸ºæŠ½çƒŸå’Œè‡ªæˆ‘æ§åˆ¶ä¹‹é—´çš„å…³ç³»å»ºç«‹ä¸€ä¸ªæ­£æ€å›å½’æ¨¡å‹ï¼š

$$\begin{equation}
\begin{array}{lcrl} 
\text{data:} & \hspace{.05in} & Y_i|\beta_0,\beta_1,\sigma & \stackrel{ind}{\sim} N(\mu_i, \; \sigma^2) \;\; \text{ with } \;\; \mu_i = \beta_0 + \beta_1 X_{i2}\\
\text{priors:} & & \beta_{0}  &  \sim N\left(60, 2^2\right)  \\
                     & & \beta_1  & \sim N\left(-10, 4^2 \right) \\
                     & & \sigma & \sim \text{Exp}(0.6)  .\\
\end{array}
\end{equation}
$$

**åœ¨è¿™ä¸ªä¾‹å­ä¸­ï¼Œå„å‚æ•°çš„æ„ä¹‰ï¼š**

* $\beta_0  =  \beta_0 + \beta_1 \cdot 0$

    * å½“$X_{i2} = 0$æ—¶ï¼Œ$mu_i$çš„å–å€¼ã€‚ $\beta_0 $ è¡¨ç¤ºä¸æŠ½çƒŸæƒ…å†µä¸‹ï¼Œè‡ªæˆ‘æ§åˆ¶åˆ†æ•°çš„å‡å€¼

* $\beta_0 + \beta_1 \cdot 1 = \beta_0 + \beta_1 $

    * $\beta_1 $ ï¼šå½“$X_{i2} = 1$æ—¶ï¼Œ$mu_i$çš„å–å€¼ã€‚è¡¨ç¤ºæŠ½çƒŸæƒ…å†µä¸‹ï¼Œè‡ªæˆ‘æ§åˆ¶åˆ†æ•°çš„å‡å€¼ï¼Œåˆ™$\beta_1 $è¡¨ç¤ºäº†ä»ä¸æŠ½çƒŸåˆ°æŠ½çƒŸï¼Œè‡ªæˆ‘æ§åˆ¶åˆ†æ•°çš„å˜åŒ–ç¨‹åº¦

* è¿™é‡Œæˆ‘ä»¬è¿˜æ˜¯æ ¹æ®ç»éªŒå®šä¹‰å…ˆéªŒä¸­çš„è¶…å‚

**æ¥ä¸‹æ¥æˆ‘ä»¬ç›´æ¥è¿è¡Œæ¨¡å‹å®šä¹‰ã€MCMCé‡‡æ ·ã€åéªŒé¢„æµ‹éƒ¨åˆ†**

```python
with pm.Model() as category_model:

    beta_0 = pm.Normal("beta_0", mu=60, sigma=2)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=-10, sigma=4)         #å®šä¹‰beta_1
    sigma = pm.Exponential("sigma", 0.6)                  #å®šä¹‰sigma

    smoke = pm.MutableData("smoke",df_METU.smoke, dims="obs_id")
    mu = pm.Deterministic("mu", beta_0 + beta_1*smoke, dims="obs_id")                           #å®šä¹‰muï¼Œè®²è‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ

    likelihood = pm.Normal("y_est", mu=mu, sigma=sigma, observed=df_METU.scontrol, dims="obs_id")   #å®šä¹‰ä¼¼ç„¶ï¼šé¢„æµ‹å€¼yç¬¦åˆN(mu, sigma)åˆ†å¸ƒ
                                                                                #é€šè¿‡ observed ä¼ å…¥å®é™…æ•°æ®y è‡ªæˆ‘æ§åˆ¶æ°´å¹³
```

```python
#===========================
#     æ³¨æ„ï¼ï¼ï¼ä»¥ä¸‹ä»£ç å¯èƒ½éœ€è¦è¿è¡Œ1-2åˆ†é’Ÿå·¦å³
#===========================
with category_model:
    cat_regression = pm.sample(draws=5000,                   # ä½¿ç”¨mcmcæ–¹æ³•è¿›è¡Œé‡‡æ ·ï¼Œdrawsä¸ºé‡‡æ ·æ¬¡æ•°
                      tune=1000,                    # tuneä¸ºè°ƒæ•´é‡‡æ ·ç­–ç•¥çš„æ¬¡æ•°ï¼Œå¯ä»¥å†³å®šè¿™äº›ç»“æœæ˜¯å¦è¦è¢«ä¿ç•™
                      chains=4,                     # é“¾æ•°
                      discard_tuned_samples= True,  # tuneçš„ç»“æœå°†åœ¨é‡‡æ ·ç»“æŸåè¢«ä¸¢å¼ƒ
                      random_seed=84735)
```
