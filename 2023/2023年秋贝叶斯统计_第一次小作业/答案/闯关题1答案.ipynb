## STEP2ï¼šç”Ÿæˆä¸åŒæ¡ä»¶ä¸‹çš„ååº”æ—¶  
1. ç”Ÿæˆ50ä¸ªæ¡ä»¶Aä¸‹çš„ååº”æ—¶ï¼Œæ¡ä»¶Aä¸‹çš„ååº”æ—¶æœä»å‡å€¼ä¸º350ï¼Œæ ‡å‡†å·®ä¸º25çš„æ­£æ€åˆ†å¸ƒï¼Œå˜é‡åå­˜ä¸ºâ€œrt_Aâ€  
2. ç”Ÿæˆ50ä¸ªæ¡ä»¶Bä¸‹çš„ååº”æ—¶ï¼Œæ¡ä»¶Bä¸‹çš„ååº”æ—¶æœä»å‡å€¼ä¸º450ï¼Œæ ‡å‡†å·®ä¸º35çš„æ­£æ€åˆ†å¸ƒï¼Œå˜é‡åå­˜ä¸ºâ€œrt_Bâ€  
3. è®¡ç®—æ¡ä»¶Aä¸æ¡ä»¶Bååº”æ—¶çš„å·®å€¼(æ¡ä»¶Bå‡å»æ¡ä»¶A)ï¼Œå˜é‡åå­˜ä¸ºâ€œrt_diffâ€  
* æ³¨ï¼šå°†æ‰€æœ‰ç”Ÿæˆçš„ååº”æ—¶å­˜ä¸ºæ•´æ•°

```python
rt_A = np.random.normal(loc = 350,
                 scale = 25,
                 size = 50)
rt_A = rt_A.astype(int)
```

```python
rt_B = np.random.normal(loc = 450,
                 scale = 35,
                 size = 50)
rt_B = rt_B.astype(int)
```

```python
rt_diff = rt_B - rt_A
rt_diff = rt_diff.astype(int)
```

```python
df_rt = pd.DataFrame({'sub_ID': sub_array,
                      'rt_A': rt_A, 
                      'rt_B': rt_B, 
                      'rt_diff': rt_diff})
```
