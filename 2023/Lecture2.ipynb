#### ä¼¼ç„¶(likelihood)  

**ä¼¼ç„¶çš„å®šä¹‰**  

å›åˆ°æ„Ÿå¹å·æ–°é—»æ ‡é¢˜çš„ä¾‹å­ï¼Œæˆ‘ä»¬æ‰€è·å¾—çš„æ•°æ®(data)åªæ˜¯å‘Šè¯‰æˆ‘ä»¬è¿™ç¯‡æ–‡ç« çš„æ ‡é¢˜å‡ºç°äº†æ„Ÿå¹å·ï¼Œè€Œæ ¹æ®æ¡ä»¶æ¦‚ç‡çš„ä¿¡æ¯å¯çŸ¥ï¼Œ$P(A|B) = 0.2667ï¼ŒP(A|B^{c}) = 0.0222$ï¼Œæ ‡é¢˜å«æœ‰æ„Ÿå¹å·çš„æ–‡ç« æ›´å¯èƒ½(more likely)æ˜¯å‡æ–°é—»ã€‚

è¿™å°±æ˜¯ä¼¼ç„¶å‡½æ•°(likelihood function)çš„å«ä¹‰ï¼Œå®ƒè¡¨æ˜äº†æ•°æ®Aåœ¨å„ç§å‡è®¾ä¸‹å‡ºç°çš„æ¦‚ç‡ï¼Œå®ƒæ˜¯ä¸€ç³»åˆ—æ¡ä»¶æ¦‚ç‡çš„ç»„åˆ

è¿™è¡¨æ˜äº†æ ‡é¢˜å«æœ‰æ„Ÿå¹å·åœ¨â€œå‡æ–°é—»â€ä¸â€œäº‹å®æŠ¥é“â€è¿™ä¸¤ç§å‡è®¾ä¸‹å‡ºç°çš„æ¦‚ç‡  

*æ³¨æ„ï¼Œåœ¨ä¼¼ç„¶å‡½æ•°ä¸­ï¼Œæ•°æ®æ˜¯å·²çŸ¥å‘ç”Ÿçš„ï¼Œè€Œå‡è®¾æ˜¯å¯èƒ½å‘ç”Ÿçš„  


### æ¦‚ç‡(Probability) vs ä¼¼ç„¶(likelihood)  

å½“å‡è®¾$B$æ˜¯å·²çŸ¥çš„ï¼Œæ¡ä»¶æ¦‚ç‡å‡½æ•°$P(Â·|B)$å…è®¸æˆ‘ä»¬æ¯”è¾ƒåœ¨å‡è®¾$B$å‘ç”Ÿçš„å‰æä¸‹ï¼Œå„ç§å¯èƒ½äº‹ä»¶$A$æˆ–$A^{c}$å‘ç”Ÿçš„æ¦‚ç‡ã€‚

### åˆ†æ¯ï¼ˆnormalizing constantï¼‰-- **è¾¹é™…æ¦‚ç‡ (marginal probability)**  

ä¼¼ç„¶å‡½æ•°æœ‰æ•ˆæ€»ç»“äº†åœ¨çœŸå‡æ–°é—»ä¸­çš„æ„Ÿå¹å·ä½¿ç”¨ï¼Œä½†æˆ‘ä»¬ä»ç„¶éœ€è¦å¾—çŸ¥ï¼Œåœ¨æ‰€æœ‰æ–°é—»æ ‡é¢˜ä¸­ä½¿ç”¨æ„Ÿå¹å·çš„æ¦‚ç‡å¦‚ä½•, è¿™è¢«ç§°ä¸ºè¾¹é™…æ¦‚ç‡ $P(A)$  


$$  
Posterior \sim P(B|A) = \frac {data * prior}{Average probability of data} = \frac{P(A\cap B)}{P(A)} = \frac{L(B|A) * P(B)}{P(A)}  
$$  
 
æ‰€æœ‰æ–°é—»æ ‡é¢˜ä¸­ä½¿ç”¨æƒŠå¹å·çš„æ¦‚ç‡å¯è¢«åˆ†ä¸ºä¸¤éƒ¨åˆ†ï¼š

2. äº‹å®æŠ¥é“çš„æ ‡é¢˜å‡ºç°æ„Ÿå¹å·$P(A\cap B^{c})$  

ä¸¤è€…ä¹‹å’Œå³ä¸ºè¾¹é™…æ¦‚ç‡ã€‚  
- $$  
 P(A) = P(A \cap B) + P(A \cap B^{c}) = L(B|A) * P(B) + L(B^{c} | A) * P(B^{c})  
 $$  
- $$ P(A) = 0.2667 * 0.4 + 0.0222 * 0.6 = 0.12  
 $$  


### åéªŒæ¦‚ç‡æ¨¡å‹(Posterior probability model via Bayesâ€™ Rule)

**ç›´è§‰ç†è§£**  

ç°åœ¨ï¼Œæˆ‘ä»¬æ¥å°è¯•è®¡ç®—åéªŒæ¦‚ç‡ï¼Œæ¯”å¦‚è®¡ç®—æ ‡é¢˜ä½¿ç”¨æ„Ÿå¹å·çš„æ–‡ç« æ˜¯å‡æ–°é—»çš„æ¦‚ç‡$P(B|A)$ã€‚  

åœ¨æ­£å¼è®¡ç®—ä¹‹å‰ï¼Œæˆ‘ä»¬å¯ä»¥å›é¡¾ä¸€ä¸‹è¿™ä¸ªè¡¨æ ¼æ¥å»ºç«‹ä¸€äº›ç›´è§‰ã€‚

|Total|0.4|0.6|1|  

</center>  

å› ä¸ºè¿™ç¯‡æ–‡ç« æ ‡é¢˜ä½¿ç”¨äº†æ„Ÿå¹å·ï¼Œæ‰€ä»¥æˆ‘ä»¬ç›´æ¥é”å®šAè¡Œï¼Œ  
- åœ¨Aè¡Œä¸­ï¼Œ88.9%(0.1067/0.12)çš„æ–‡ç« ä¸ºå‡æ–°é—»ï¼Œ11.1%(0.0133/0.12)çš„æ–‡ç« ä¸ºäº‹å®ã€‚  
- å› æ­¤æœ‰88.9%çš„åéªŒæ¦‚ç‡è®¤ä¸ºå½“å‰è¿™ä¸€æ–‡ç« å±äºå‡æ–°é—»ã€‚  


**æ­£å¼è®¡ç®—**  

å¦‚ä½•å‡­å€Ÿè´å¶æ–¯å…¬å¼çš„æ•°å­¦å½¢å¼æ¨å¯¼å¾—åˆ°è¯¥ç»“æœï¼Ÿ  

$$  
Posterior \sim P(B|A) = \frac {data * prior}{Average \, probability \, of \, data} ={\frac{P(A\cap B)}{P(A)}}={\frac{L(B|A) * P(B)}{L(B|A) * P(B) + L(B^{c}|A) * P(B^{c})}}  
$$  

- $P(B|A)={\frac{P(B)L(B|A)}{P(A)}}={\frac{0.4\cdot0.2667}{0.12}}=0.889$  
- å½“å¸¦å…¥ä¹‹å‰è®¡ç®—å¾—åˆ°çš„æ•°å€¼åˆ°è´å¶æ–¯å…¬å¼ä¸­ï¼Œæˆ‘ä»¬å¾—åˆ°äº†æ ‡é¢˜å¸¦æœ‰æƒŠå¹å·çš„æ–°é—»ä¸ºè™šå‡æ–°é—»çš„æ¦‚ç‡ã€‚

- å¯ä»¥æ³¨æ„åˆ°ï¼šå…ˆéªŒæ¦‚ç‡å’ŒåéªŒæ¦‚ç‡çš„å’Œå‡ç­‰äº1ã€‚

### åéªŒæ¨¡æ‹Ÿ(Posterior simulation)  

å¦‚ä½•ä½¿ç”¨ä»£ç å®ç°åéªŒæ¦‚ç‡çš„è®¡ç®—ï¼Ÿ

1. å®šä¹‰æ–‡ç« ç±»å‹ä¸ç›¸åº”çš„å…ˆéªŒæ¦‚ç‡

```python
# å®šä¹‰æ–‡ç« ç±»å‹
article = pd.DataFrame({'type': ['real', 'fake']})

# å®šä¹‰å…ˆéªŒæ¦‚ç‡
prior = [0.6, 0.4]
```

2. æ¨¡æ‹Ÿä¸€äº›å¯èƒ½è¢«æŠ•æ”¾ç»™ä½ çš„æ–‡ç« 

```python
# æ¨¡æ‹Ÿç”Ÿæˆ 10000 æ–‡ç« ï¼ŒåŒ…æ‹¬å…¶ç±»å‹
np.random.seed(84735)
article_sim = article.sample(n=10000, weights=prior, replace=True)
# æŸ¥çœ‹å‰ 10 è¡Œæ•°æ®
article_sim.head(10)
```

```python
#æˆ‘ä»¬å¯ä»¥é€šè¿‡ç”»å›¾æ¥æŸ¥çœ‹è¿™äº›è¢«æŠ•æ”¾çš„æ–‡ç« ä¸­ï¼Œäº‹å®æŠ¥é“å’Œå‡æ–°é—»çš„æ¯”ä¾‹ã€‚
article_sim['type'].value_counts().plot.bar()
plt.xticks(rotation=0)
plt.show()
```

3. æ¥ä¸‹æ¥æˆ‘ä»¬éœ€è¦æ¨¡æ‹Ÿ10000ç¯‡æ–°é—»æ ‡é¢˜çš„æ„Ÿå¹å·ä½¿ç”¨æƒ…å†µï¼Œ  
- å’Œä¹‹å‰ç›¸åŒï¼Œå‡æ–°é—»ä½¿ç”¨æ„Ÿå¹å·çš„å¯èƒ½æ€§ä¸º26.67% ï¼Œ  
- äº‹å®æŠ¥é“ä½¿ç”¨æ„Ÿå¹å·çš„å¯èƒ½æ€§ä¸º2.22% 

```python
article_sim['data_model'] = np.where(article_sim['type'] == 'fake', 0.2667, 0.0222)

# å®šä¹‰æ–°é—»æ ‡é¢˜æ˜¯å¦ä½¿ç”¨æƒŠå¹å·
data = ['yes', 'no']

# è®¾ç½®éšæœºç§å­ï¼Œä»¥ä¾¿å¾—åˆ°é‡å¤çš„ç»“æœ
rng=np.random.default_rng(84735)
# ç”ŸæˆæƒŠå¹å·ç›¸å…³çš„æ•°æ®
article_sim['usage'] = article_sim.apply(lambda x: rng.choice(data, 1, p = [x.data_model, 1-x.data_model])[0], axis=1)
```

```python
# æ˜¾ç¤ºæ¯ä¸ªç±»åˆ«æ–°é—»æ•°é‡
(
  article_sim.groupby(['usage', 'type'])
    .size()
    .unstack(fill_value=0)
)
```

- å¤§çº¦26%(1039/(1039+2899))çš„å‡æ–°é—»æ ‡é¢˜ä½¿ç”¨äº†æ„Ÿå¹å·, $P(A|B)=0.26$  

ç»“åˆå…ˆéªŒå’Œä¼¼ç„¶ï¼Œæˆ‘ä»¬å¯ä»¥è¿›ä¸€æ­¥è®¡ç®—åˆ†æ¯(è¾¹é™…æ¦‚ç‡)ï¼š  
- $L(B|A)*P(B) + L(B^{c}|A)*P(B^{c}) = 0.26*0.4 + 0.02*0.6 = 0.116$  

æœ€åï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®—çš„åˆ°åéªŒ (å¸¦æœ‰æ„Ÿå¹æ ‡é¢˜çš„æ–°é—»ä¸­ï¼Œè™šå‡æ–°é—»çš„æ¦‚ç‡)ï¼š

- åœ¨10000ç¯‡æ–°é—»ä¸­ï¼Œæ ‡é¢˜ä½¿ç”¨æ„Ÿå¹å·æ–°é—»æœ‰1164ç¯‡(åˆ†æ¯)  
- è€Œç°åœ¨ï¼Œæˆ‘ä»¬å¯ä»¥çŸ¥é“ï¼Œåœ¨ä½¿ç”¨æ„Ÿå¹å·æ ‡é¢˜çš„æ–‡ç« ä¸­ï¼Œ89%(1039/1164)çš„æ–‡ç« ä¸ºå‡æ–°é—»

### å…ˆéªŒæ¦‚ç‡æ¨¡å‹(**Prior** probability model)

åƒå‡æ–°é—»åˆ†æä¸­ä¸€æ ·ï¼Œæˆ‘ä»¬å¯ä»¥ç»™è¿™äº›éšæœºå˜é‡$\pi$å»ºç«‹å…ˆéªŒæ¨¡å‹

2. ä¸ºæ¯ä¸ªå€¼åˆ†é…ä¸€ä¸ªå…ˆéªŒæ¦‚ç‡  
3. è¿™äº›æ¦‚ç‡ä¹‹å’Œä¸º1  

<center>  

| $\pi$	    |0.2  |0.5 |0.8 |Total  
|---------- |-----|----|----|-----|  
|$f(\pi)$   |0.10  |0.25 |0.65   |1|  

</center>  

$\pi$è¡¨ç¤ºKasparovçš„èƒœç‡å¯èƒ½ä¸º20%ï¼Œ50%ï¼Œ80%ï¼Œ$f(\pi)$åˆ™è¡¨ç¤ºæ¯ä¸ªå¯èƒ½$\pi$å€¼çš„å…ˆéªŒæ¦‚ç‡

æ³¨ï¼š  

* è¿™é‡Œçš„æ¨¡å‹å¯¹ç°å®è¿›è¡Œäº†ç®€åŒ–ï¼ŒæŒ‰ç†æ¥è¯´æ£‹æ‰‹çš„èƒœç‡å¯ä»¥å–0-1ä¸­çš„ä»»ä½•å€¼ï¼Œè¿™é‡Œæˆ‘ä»¬åªä½¿ç”¨äº†ä¸‰ä¸ªå€¼ 0.2ï¼Œ0.5ï¼Œ0.8  

* è¡¨æ ¼ä¸­çš„ç¬¬äºŒè¡Œæ˜¾ç¤ºäº†æ¯ä¸ªå¯èƒ½å–å€¼$\pi$çš„å…ˆéªŒæ¦‚ç‡ï¼Œè¿™åˆè¢«ç§°ä¸ºæ¦‚ç‡è´¨é‡å‡½æ•°(probability mass function (pmf))


**æ¦‚ç‡è´¨é‡å¯†åº¦å‡½æ•°(probability mass function, pmf)ï¼š** ç”¨æ¥æè¿°ç¦»æ•£å‹éšæœºå˜é‡åœ¨å„ç‰¹å®šå–å€¼ä¸Šçš„æ¦‚ç‡  

å¯¹äºç¦»æ•£å‹éšæœºå˜é‡$Y$ï¼Œ$Y$å„å–å€¼çš„æ¦‚ç‡ç”±$f(y)$æŒ‡å®šï¼š  
$$  
f(y) = P(Y=y)  
$$  

å¹¶ä¸”æœ‰å¦‚ä¸‹æ€§è´¨ï¼š  

* å¯¹æ‰€æœ‰yçš„å–å€¼æ¥è¯´ï¼Œ$0\leq f(y) \leq 1$  
* $\sum_{all\,\pmb{y}}f(y) = 1$ï¼Œyå–å€¼çš„æ‰€æœ‰æ¦‚ç‡ä¹‹å’Œä¸º1

### äºŒé¡¹å¼æ•°æ®æ¨¡å‹(The Binomial **data** model)

**æ”¶é›†æ•°æ®**  

ç¬¬äºŒæ­¥ï¼Œæˆ‘ä»¬éœ€è¦æ”¶é›†æ¯”èµ›çš„æ•°æ®ç»“æœã€‚æˆ‘ä»¬çš„æ•°æ®$Y$æ˜¯Kasparovåœ¨1997å¹´çš„6åœºæ¯”èµ›ä¸­çš„è·èƒœæ¬¡æ•°ã€‚  

$Y$å¯ä»¥å–{0,1,...,6}ä¸­çš„ä»»æ„å€¼ï¼Œå¹¶ä¸”$Y$çš„å–å€¼å—åˆ°$\pi$çš„å½±å“ï¼Œè¯•æƒ³å¦‚æœ$\pi$ä¸º0.8ï¼Œæ„å‘³ç€Kasparovçš„èƒœç‡å¾ˆé«˜ï¼Œé‚£ä¹ˆ$Y$çš„å–å€¼ä¹Ÿä¼šåé«˜ã€‚æˆ‘ä»¬åŒæ ·å¯ä»¥ç”¨æ¡ä»¶æ¦‚ç‡å…¬å¼æ¥è¡¨æ˜è¿™ä¸€ç‚¹ï¼š  

$$  
f(y|\pi) = P(Y=y|\pi)  
$$  

* å¯¹æ‰€æœ‰yçš„å–å€¼æ¥è¯´ï¼Œ$0\leq f(y|\pi) \leq 1$  
* $\sum_{all\,\pmb{y}}f(y|\pi) = 1$  
--------------------------------------------------------------------  

æˆ‘ä»¬æå‡ºä»¥ä¸‹ä¸¤ä¸ªå‡è®¾ï¼š  

ï¼ˆ1ï¼‰ä»»ä½•æ¯”èµ›çš„ç»“æœéƒ½ç›¸äº’ç‹¬ç«‹ã€äº’ä¸å½±å“  

ï¼ˆ2ï¼‰Kasparovåœ¨æ¯åœºæ¯”èµ›ä¸­çš„èƒœç‡éƒ½æ˜¯ç›¸åŒçš„  

åœ¨è¿™ä¸¤ä¸ªå‰æä¸‹ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨**äºŒé¡¹æ¨¡å‹**æ¥å¯¹Kasparovçš„æ¯”èµ›ç»“æœè¿›è¡Œæ‹Ÿåˆ

**äºŒé¡¹æ¨¡å‹**  

$$  
f(y|\pi) = \binom{n}{y} \pi^{y}(1-\pi)^{n-y} \quad\quad for\;y \in \{0,1,2,...,n\}  
$$  
$$  
\binom{n}{y} = \frac{n!}{y!(n-y)!}  
$$  

$Y$è¡¨ç¤ºåœ¨$n$ä¸ªè¯•æ¬¡ä¸­æˆåŠŸçš„æ¬¡æ•°ï¼ŒäºŒé¡¹æ¨¡å‹å«æœ‰çš„å‰æå‡è®¾æ˜¯ï¼š  

(1) æ‰€æœ‰è¯•æ¬¡å‘ç”Ÿéƒ½æ˜¯ç›¸äº’ç‹¬ç«‹çš„  

(2) åœ¨æ¯ä¸ªè¯•æ¬¡ä¸­ï¼ŒæˆåŠŸçš„æ¦‚ç‡éƒ½æ˜¯ä¸€ä¸ªå›ºå®šçš„å€¼$\pi$

**ä½¿ç”¨äºŒé¡¹æ¨¡å‹æ¥æè¿°è·èƒœæ¬¡æ•°ä¸èƒœç‡çš„å…³ç³»**  

åœ¨$n = 6$çš„æ¯”èµ›ä¸­ï¼ŒKasparovçš„è·èƒœæ¬¡æ•°ä¸èƒœç‡çš„å…³ç³»ç¬¦åˆäºŒé¡¹æ¨¡å‹  

$$  
Y|\pi \sim Bin(6,\pi)  
$$  

Conditional PMF  

$$  
f(y|\pi) = \binom{6}{y} \pi^{y}(1-\pi)^{6-y} \quad\quad for\;y \in \{0,1,2,3,4,5,6\}  
$$  

-----------------------------------  

**å›¾ä¾‹**  

ä¸‹å›¾ç»™å‡ºäº†å‡ ç§pmf $f(y|\pi)$ï¼Œå³åœ¨ä¸åŒèƒœç‡$\pi$ä¸‹ï¼Œä¸åŒè·èƒœæ¬¡æ•°$Y$å‘ç”Ÿçš„å¯èƒ½æ€§ã€‚  

*é»‘çº¿è¡¨ç¤ºåœ¨ç‰¹å®šèƒœç‡ä¸‹ï¼ŒY=1(èµ¢ä¸€æ¬¡)çš„ä¼¼ç„¶å€¼  

![Image Name](https://www.bayesrulesbook.com/bookdown_files/figure-html/binom-chess-1.png)

è¿™ä¸ªæ¦‚ç‡è´¨é‡å¯†åº¦å‡½æ•°æ€»ç»“äº†åœ¨ç‰¹å®šèƒœç‡$\pi$ä¸‹è§‚å¯Ÿåˆ°è·èƒœæ¬¡æ•°$Y=y$çš„æ¡ä»¶æ¦‚ç‡  

ä¾‹å¦‚ï¼Œå½“Kasparovçš„èƒœç‡ä¸º0.8æ—¶ï¼Œä»–èµ¢ä¸‹å…¨éƒ¨6åœºçš„å¯èƒ½æ€§å¤§çº¦ä¸º0.26  

$$  
f(y=0.6|\pi=0.8) = \binom{6}{6} 0.8^{6}(1-0.8)^{6-6} = 1\times0.8^{6}\times1 \approx 0.26  
$$  

è€Œè¾“æ‰å…¨éƒ¨6åœºæ¯”èµ›çš„å¯èƒ½æ€§åˆ™æ¥è¿‘ä¸º0  

$$  
f(y=0|\pi=0.8) = \binom{6}{0} 0.8^{0}(1-0.8)^{6-0} = 1\times0.2^{6}\times1 \approx 0.000064  
$$

### äºŒé¡¹ä¼¼ç„¶å‡½æ•°(The Binomial likelihood function)

**è·å¾—æ•°æ®**  

ç°åœ¨ï¼Œå¦‚æœæˆ‘ä»¬å¾—çŸ¥Kasporovåœ¨1997å¹´çš„6åœºæ¯”èµ›ä¸­åªè·èƒœäº†1æ¬¡ï¼Œæˆ‘ä»¬å¯ä»¥æ€è€ƒåœ¨ä¼—å¤šå¯èƒ½çš„èƒœç‡ä¸­ï¼Œå“ªä¸€ä¸ªæ›´å¯èƒ½äº§ç”Ÿæˆ‘ä»¬å½“å‰çš„æ•°æ®ç»“æœã€‚  

æ¢è¨€ä¹‹ï¼Œæˆ‘ä»¬æƒ³çŸ¥é“åœ¨æ¯ä¸ªå¯èƒ½çš„èƒœç‡$\pi$ä¸‹ï¼Œ6åœºæ¯”èµ›åªè·èƒœ1æ¬¡çš„ä¼¼ç„¶(likelihood)ã€‚  

![](https://www.bayesrulesbook.com/bookdown_files/figure-html/binom-chess-like-1.png)

**ä¼¼ç„¶å‡½æ•°**  

è·èƒœæ¬¡æ•°ä¸º1æ—¶çš„ä¼¼ç„¶å‡½æ•°  

$$  
L(\pi|y=1) = f(y=1|\pi) = \binom{6}{1} \pi^{1}(1-\pi)^{6-1} = 6\pi(1-\pi)^{5}  
$$  

æ¯”å¦‚å½“èƒœç‡ä¸º0.8æ—¶ï¼Œå‡ºç°è·èƒœæ¬¡æ•°ä¸º1çš„å¯èƒ½æ€§æ˜¯ï¼š  
$$  
L(\pi=0.8|y=1) = 6\times0.8\times(1-0.8)^{5}\approx 0.0015  
$$  

ä¸åŒèƒœç‡ä¸‹çš„ä¼¼ç„¶ï¼š

|---------------|-----  |----   |----|  
|$f(\pi \| y=1)$   |0.3932  |0.0938 |0.0015|  

</center>  



**æ³¨æ„ï¼š**  

ä¼¼ç„¶å‡½æ•°è¡¨ç¤ºçš„æ˜¯ï¼Œåœ¨å„ç§å¯èƒ½çš„èƒœç‡$\pi$ä¸‹,è·èƒœæ¬¡æ•°$Y=1$çš„å¯èƒ½æ€§ï¼Œæ‰€ä»¥  

(1)è¯¥ä¼¼ç„¶å‡½æ•°å…¬å¼åªå–å†³äº$\pi$  

(2)ä¼¼ç„¶å‡½æ•°çš„æ€»å’ŒåŠ èµ·æ¥ä¸ä¸º1ï¼ˆä»æ¡ä»¶æ¦‚ç‡çš„å…¬å¼æ¥çœ‹ï¼Œä¼¼ç„¶å‡½æ•°çš„åˆ†æ¯æ˜¯ä¸åŒçš„ï¼‰

**Conditional PMF VS likelihood function**  

å½“$\pi$æ˜¯å·²çŸ¥çš„ï¼Œæ¡ä»¶æ¦‚ç‡è´¨é‡å¯†åº¦å‡½æ•°$f(Â·|\pi)$å…è®¸æˆ‘ä»¬æ¯”è¾ƒåœ¨$\pi$å‘ç”Ÿçš„å‰æä¸‹ï¼Œå„ç§å¯èƒ½æ•°æ®$Y$(e.g., $y_{1},y_{2}$)å‘ç”Ÿçš„æ¦‚ç‡ã€‚  
$$

$$  

å½“$Y = y$æ˜¯å·²çŸ¥çš„ï¼Œä¼¼ç„¶å‡½æ•°$L(Â·|y)= f(y|Â·)$å…è®¸æˆ‘ä»¬æ¯”è¾ƒåœ¨å„ç§ä¸åŒçš„$\pi$(e.g., $\pi_{1},\pi_{2}$)ä¸‹ï¼Œè§‚å¯Ÿåˆ°æ•°æ®$y$çš„ç›¸å¯¹å¯èƒ½æ€§(relative likelihood)ã€‚

$$

### åˆ†æ¯(Normalizing constant)

**è®¡ç®—åˆ†æ¯**  

è´å¶æ–¯æ³•åˆ™ä¸­åŒ…å«äº†å…ˆéªŒã€ä¼¼ç„¶ä¸åˆ†æ¯ï¼Œå‰ä¸¤ä¸ªå·²ç»è€ƒè™‘è¿‡äº†ï¼Œæ¥ä¸‹æ¥æˆ‘ä»¬æ¥æ€è€ƒè¿™ä¸ªä¾‹å­ä¸­çš„åˆ†æ¯ã€‚  

åœ¨è¿™ä¸ªä¾‹å­ä¸­ï¼Œåˆ†æ¯ä¸º$f(y=1)$ï¼ŒæŒ‡åœ¨æ‰€æœ‰å¯èƒ½çš„èƒœç‡$\pi$ä¸‹ï¼ŒKasparovè·èƒœä¸€åœºçš„æ€»ä½“æ¦‚ç‡(total probability)ã€‚æ€»ä½“æ¦‚ç‡åˆ™æ˜¯è§‚å¯Ÿåˆ°$Y=1$ä¼¼ç„¶å€¼çš„æ€»å’Œï¼Œå¯ä»¥å†™ä¸ºï¼š

ä»£å…¥å…·ä½“çš„å€¼è®¡ç®—å¯å¾—ï¼š  

$$

### åéªŒæ¦‚ç‡æ¨¡å‹(Posterior probability model)

ä¸‹å›¾æ€»ç»“äº†å·²çŸ¥ä¿¡æ¯ï¼Œå¹¶ä½œå‡ºäº†é¢„æµ‹ï¼š  

å…ˆéªŒæ¨¡å‹å‘Šè¯‰æˆ‘ä»¬Kasparovæˆ˜èƒœæ·±è“çš„å¯èƒ½æ€§å¾ˆé«˜ã€‚  

ç„¶è€Œï¼Œåœ¨97å¹´çš„é‡é€¢å¯¹å†³ä¸­ï¼ŒKasparovåªè·èƒœä¸€æ¬¡çš„è¿™ä¸ªæ•°æ®å‘Šè¯‰æˆ‘ä»¬ä»–çš„èƒœç‡å¾ˆä½ã€‚  

ç»¼åˆå…ˆéªŒå’Œä¼¼ç„¶ï¼Œæ›´æ–°çš„åéªŒæ¨¡å‹å‘Šè¯‰æˆ‘ä»¬Kasparovæˆ˜èƒœæ·±è“çš„å¯èƒ½æ€§æœ€å¯èƒ½å·²ç»é™è‡³0.2ã€‚

**åéªŒå…¬å¼**  

ä¸Šå›¾æ‰€è¡¨ç¤ºçš„åéªŒå¯å†™æˆï¼š

è¡¨ç¤ºå½“Kasparovåªèµ¢ä¸€åœºæ—¶ï¼Œä»–å¯èƒ½çš„èƒœç‡$\pi$

### (Posterior shortcut)

è®¡ç®—åˆ†æ¯åªæ˜¯ä¸ºäº†è¿›è¡Œæ ‡å‡†åŒ–ï¼Œè®©åéªŒæ¦‚ç‡ä¹‹å’Œä¸º1

ä½†ä¹Ÿå¯ä»¥çœç•¥åˆ†æ¯çš„è®¡ç®—ï¼Œå› ä¸ºåˆ†æ¯æ˜¯ä¸€ä¸ªæ¦‚ç‡å€¼ï¼Œå³å¸¸æ•°ï¼Œé‚£ä¹ˆæˆ‘ä»¬å¯ä»¥è®©$f(y=1) = 1/c$æ¥ç®€åŒ–è¿™ä¸ªè¿‡ç¨‹  

åéªŒçš„è®¡ç®—å¯å†™æˆï¼š

$$

$\propto$ è¡¨ç¤ºæˆæ¯”ä¾‹ï¼Œå°½ç®¡è¿™äº›æœªç»æ ‡å‡†åŒ–çš„åéªŒæ¦‚ç‡æ€»å’Œä¸ç­‰äº1

ä½†å®ƒä»¬çš„æ¯”ä¾‹å…³ç³»å¹¶æœªæ”¹å˜(è§ä¸‹å›¾)

æ³¨æ„åˆ°å…¶å®åˆ†æ¯ä¹Ÿå°±æ˜¯æ‰€æœ‰ä¼¼ç„¶å€¼çš„æ€»å’Œï¼Œå› æ­¤åéªŒæ¦‚ç‡çš„è®¡ç®—å…¬å¼è¿˜å¯ä»¥å†™æˆï¼š

**Proportionality**  

æ—¢ç„¶$f(y)$æ˜¯ä¸€ä¸ªç”¨æ¥æ ‡å‡†åŒ–çš„å¸¸æ•°ï¼Œå®ƒå¹¶ä¸å—$\pi$çš„å½±å“ï¼Œé‚£ä¹ˆåéªŒæ¦‚ç‡è´¨é‡å‡½æ•°$f(\pi|y)$ å°±ä¸$f(\pi)$å’Œ$L(\pi|y)$æˆæ­£æ¯”

### Posterior simulation (with code)

### 1. å®šä¹‰å…ˆéªŒæ¨¡å‹  
- å®šä¹‰å¯èƒ½çš„èƒœç‡  
- å®šä¹‰èƒœç‡çš„å¯èƒ½æ€§

```python
import pandas as pd
import numpy as np

# å®šä¹‰å¯èƒ½çš„èƒœç‡
chess = pd.DataFrame({'pi':[0.2, 0.5, 0.8]})

# å®šä¹‰å…ˆéªŒæ¨¡å‹
prior = [0.10, 0.25, 0.65]
```

### 2. æ¨¡æ‹Ÿåœ¨ç‰¹å®šèƒœç‡ä¸‹ï¼Œ6åœºæ¯”èµ›ä¸­çš„è·èƒœæ¬¡æ•°  
- é‡å¤è¿™ä¸ªè¿‡ç¨‹10000æ¬¡

```python
# è®¾ç½®éšæœºæ•°ç§å­ä¿è¯å¯é‡å¤æ€§
np.random.seed(84735)

# ä»å…ˆéªŒä¸­æŠ½å–10000ä¸ª pi å€¼ï¼Œå¹¶ç”Ÿæˆå¯¹åº”çš„yå€¼

chess_sim = chess.sample(n=10000, weights=prior, replace=True)

chess_sim['y'] = np.random.binomial(n=6, p=chess_sim['pi'], size=len(chess_sim))

chess_sim.head(10)
```

```python
#å¯¹piçš„æŠ½å–æƒ…å†µè¿›è¡Œæ€»ç»“
chess_counts =  chess_sim['pi'].value_counts().reset_index()

chess_counts.columns = ['pi','n']

chess_counts['percentage'] = (chess_counts['n']/len(chess_sim))

chess_counts = chess_counts.sort_values(by='pi')

print(chess_counts)
```

### 3.  ä¸åŒèƒœç‡ä¸‹ï¼Œä¸åŒè·èƒœæ¬¡æ•°çš„åˆ†å¸ƒæƒ…å†µ$f(y|\pi)$

```python
# å¯¼å…¥ç»˜å›¾å·¥å…· seaborn
import seaborn as sns
chess_lik = sns.FacetGrid(chess_sim,col="pi")
chess_lik.map(sns.histplot,'y',stat='probability',discrete=True)
```

### 4. æŸ¥çœ‹$y=1$æ—¶ï¼Œå¯¹åº”çš„$\pi$çš„åˆ†å¸ƒæƒ…å†µ

```python
chess_post = chess_sim[chess_sim['y'] == 1].value_counts()
chess_post
```
