### A Beta-Binomial example  

ç°åœ¨å‡è®¾å…ˆéªŒ$\pi$ æœä»  

$$  
\pi    \sim \text{Beta}(2, 2)  
$$  

ä¼¼ç„¶å‡½æ•°ä¸ºï¼š  
$$  
Y|\pi  \sim \text{Bin}(10, \pi)  
$$  

> * è¿™æ˜¯æˆ‘ä»¬åœ¨lec3ç†Ÿæ‚‰çš„beta-binomialæ¨¡å‹ï¼Œå‡è®¾$\pi$åæ˜ çš„æ˜¯æ”¯æŒç‡ï¼Œä¼¼ç„¶å‡½æ•°åæ˜ çš„åˆ™æ˜¯åœ¨æŸä¸ªæ”¯æŒç‡ä¸‹ï¼Œæ€»æŠ•ç¥¨æ•°ä¸º10æ—¶ï¼Œæ”¯æŒç¥¨æ•°çš„åˆ†å¸ƒæ¦‚ç‡æƒ…å†µã€‚  
> $Y | \pi  \sim \text{Bin}(n, \pi)$  
> $\pi \sim \text{Beta}(\alpha, \beta)$  
> * åœ¨è§‚å¯Ÿåˆ° n æ¬¡æ—¶é—´ä¸­æœ‰ Y = yæ¬¡ç›®æ ‡äº‹ä»¶åï¼Œ$\pi$çš„åéªŒåˆ†å¸ƒå¯ä»¥ç”¨Betaæ¨¡å‹æ¥æè¿°ï¼Œåæ˜ äº†å…ˆéªŒï¼ˆé€šè¿‡Î±å’ŒÎ²ï¼‰å’Œæ•°æ®ï¼ˆé€šè¿‡yå’Œnï¼‰çš„å½±å“ï¼š  
> $\pi | (Y = y) \sim \text{Beta}(\alpha + y, \beta + n - y) $  


å‡è®¾æ”¯æŒç¥¨æ•°ä¸º9ï¼Œæˆ‘ä»¬é€šè¿‡å…±è½­å…ˆéªŒçš„å…¬å¼ï¼Œå¯ä»¥ç›´æ¥è®¡ç®—å‡º$Beta$åéªŒåˆ†å¸ƒçš„ä¸¤ä¸ªå‚æ•°ï¼š  
- $Y + \alpha = 9 + 2\;\;, \;\;\;\;\;\;\;\; n - Y + \beta = 10 - 9 + 2$  
- $\pi | (Y = 9) \sim \text{Beta}(11, 3)$

**ä»£ç ç¤ºä¾‹ï¼šç½‘æ ¼è¿‘ä¼¼ä¼°è®¡åéªŒåˆ†å¸ƒ**  

ç°åœ¨æˆ‘ä»¬æš‚æ—¶å¿˜è®°åéªŒçš„ç®€ä¾¿è®¡ç®—æ–¹æ³•ï¼Œä½¿ç”¨ç½‘æ ¼è¿‘ä¼¼æ¥å¯¹åéªŒåˆ†å¸ƒè¿›è¡Œä¼°è®¡ã€‚  

**step1ï¼š** é¦–å…ˆï¼Œæˆ‘ä»¬å°†ä»0~1å†…çš„è¿ç»­å˜é‡$\pi$ä¸­å–å‡º6ä¸ªå€¼ï¼š$\pi \in \{0, 0.2, 0.4, 0.6, 0.8, 1\}$

````python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
pi_grid = np.linspace(0,1,6)
print("ä»0~1å†…çš„è¿ç»­å˜é‡Ï€ä¸­å–å‡º6ä¸ªå€¼ï¼š", pi_grid)
````

**step2ã€3ï¼š**  
    åœ¨æ¯ä¸€ä¸ª$\pi$ä¸‹ï¼Œè®¡ç®—å…ˆéªŒåˆ†å¸ƒ $Beta(2, 2)$ï¼Œä¸ä¼¼ç„¶å‡½æ•°$Bin(10, \pi)$ (Y=9)çš„ä¹˜ç§¯ï¼Œè®¡ç®—æ€»å’Œå¹¶è¿›è¡Œå½’ä¸€åŒ–

````python
prior = st.beta.pdf(pi_grid, 2, 2)
likelihood = st.binom.pmf(9, 10, pi_grid)
posterior = prior * likelihood / np.sum(prior * likelihood)

plt.stem(pi_grid, posterior,    # ä½¿ç”¨ plt.stem() ç»˜åˆ¶å‚ç›´æŸ±çŠ¶å›¾ï¼Œè¡¨ç¤ºåœ¨ pi_grid ä¸Šçš„ posterior åˆ†å¸ƒ
         linefmt='black',
         bottom=-1)  

plt.ylim(0, 1)                  # è®¾ç½® y è½´èŒƒå›´åœ¨ 0 åˆ° 1 ä¹‹é—´
plt.xlabel('pi_grid')           # è®¾ç½® y è½´èŒƒå›´åœ¨ 0 åˆ° 1 ä¹‹é—´
plt.ylabel('Posterior')         # è®¾ç½® y è½´èŒƒå›´åœ¨ 0 åˆ° 1 ä¹‹é—´
sns.despine()                   # ä½¿ç”¨ sns.despine() å‡½æ•°å»é™¤å›¾å½¢çš„ä¸Šæ–¹å’Œå³ä¾§çš„è¾¹æ¡†
````

**step4ï¼š**  
    å¾—åˆ°åéªŒç»“æœ($\pi$)çš„åˆ†å¸ƒåï¼Œä»è¿™ä¸ªåˆ†å¸ƒä¸­æŠ½æ ·10000æ¬¡  

*ç”±äº$\pi$çš„å–å€¼åªåœ¨ $\{0, 0.2, 0.4, 0.6, 0.8, 1\}$ä¹‹ä¸­ï¼Œä»ä¸Šå›¾ä¸­å¯ä»¥çœ‹åˆ°ï¼Œåªæœ‰0.6ä¸0.8çš„æ¦‚ç‡å€¼ç›¸å¯¹è¾ƒé«˜

````python
np.random.seed(84735)

posterior_sample = np.random.choice(pi_grid,        # pi_grid ä¸ºæŠ½æ ·çš„å€™é€‰é›†åˆ
                                    size=10000,     # ä» posterior åˆ†å¸ƒä¸­éšæœºæŠ½å– 10000 ä¸ªæ ·æœ¬
                                    p=posterior,    # p=posterior ä¸ºæŠ½æ ·çš„æ¦‚ç‡åˆ†å¸ƒ
                                    replace=True)   # replace=True è¡¨ç¤ºæœ‰æ”¾å›æŠ½æ ·

posterior_sample = pd.DataFrame({"pi_sample": posterior_sample})    # å°†æŠ½å–çš„æ ·æœ¬å­˜å‚¨åœ¨ DataFrame ä¸­ï¼Œåˆ—åä¸º "pi_sample"

posterior_sample.value_counts(normalize=True).reset_index()        # å¯¹ posterior_sample ä¸­çš„æ ·æœ¬è¿›è¡Œè®¡æ•°ï¼Œå¹¶ä½¿ç”¨ normalize=True å°†è®¡æ•°è½¬æ¢ä¸ºç›¸å¯¹é¢‘ç‡
                                                                   # ç„¶åé€šè¿‡ reset_index() å°†ç»“æœé‡æ–°è®¾ç½®ä¸º DataFrame
````

**æŠ½æ ·ç»“æœå›¾ç¤º**  

å¯ä»¥çœ‹åˆ°ï¼Œåœ¨æŠ½æ ·çš„ç»“æœä¸­ï¼Œ$\pi$å¤§å¤šéƒ½æ˜¯0.6æˆ–0.8ï¼Œå°‘éƒ¨åˆ†æ˜¯0.4  

æˆ‘ä»¬å¯¹æ¯”ä¸€ä¸‹æŠ½æ ·å¾—åˆ°çš„åéªŒåˆ†å¸ƒå›¾ å’Œ å®é™…çš„åéªŒåˆ†å¸ƒ$Beta(11, 3)$

````python
x_beta = np.linspace(0, 1, 10000)       # ç”Ÿæˆ10000 ä¸ªç‚¹ï¼ŒèŒƒå›´åœ¨ [0, 1] ä¹‹é—´

y_beta = st.beta.pdf(x_beta, 11, 3)     # ç”Ÿæˆ Beta(11,3)

plt.plot(x_beta, y_beta, color='black') # ç»˜åˆ¶ beta(11,3)


plt.hist(posterior_sample["pi_sample"], # ç»˜åˆ¶ 10000æ¬¡æŠ½æ ·ç»“æœ
         edgecolor = "black",
         color="grey",
         alpha = 0.7,
         density=True,)

sns.despine()
````

**å¢åŠ ç½‘æ ¼æ•°**  

* è¿™ä¸ªç»“æœå½“ç„¶è¿‡åº¦ç®€åŒ–äº†åéªŒåˆ†å¸ƒï¼Œè¿˜è®°å¾—æˆ‘ä»¬åˆšåˆšè¯´ï¼Œåªè¦å–çš„ç½‘æ ¼è¶Šå¤šï¼Œå¯¹åéªŒçš„ä¼°è®¡ä¼šæ›´å‡†ç¡®ã€‚  

* é‚£ä¹ˆï¼Œç°åœ¨$\pi$ä¸å–6ä¸ªå€¼ï¼Œè€Œæ˜¯åœ¨0~1ä¹‹é—´å–101ä¸ªå€¼$\pi \in \{0, 0.01, 0.02, \ldots, 0.99, 1\}$  

* åŒæ ·çš„ï¼Œæˆ‘ä»¬æ‰§è¡Œä¸Šè¿°å››ä¸ªæ­¥éª¤ï¼š

````python
pi_grid = np.linspace(0, 1, 101)            # ç”Ÿæˆä¸€ä¸ª 101 ä¸ªç‚¹ï¼ŒèŒƒå›´åœ¨ [0, 1] ä¹‹é—´

prior = st.beta.pdf(pi_grid, 2, 2)          # ç”Ÿæˆ Beta(2,2)


likelihood = st.binom.pmf(9, 10, pi_grid)   # ç”ŸæˆäºŒé¡¹åˆ†å¸ƒ, å‚æ•°ä¸º n=10ï¼ˆæ€»è¯•éªŒæ¬¡æ•°ï¼‰ï¼Œk=9ï¼ˆæˆåŠŸæ¬¡æ•°ï¼‰ï¼Œä»¥åŠ pi_grid ä¸­çš„æ¯ä¸ªæ¦‚ç‡å€¼


posterior = prior * likelihood / np.sum(prior * likelihood)     # è®¡ç®—åéªŒæ¦‚ç‡ï¼Œå³å…ˆéªŒæ¦‚ç‡å’Œä¼¼ç„¶å‡½æ•°çš„ä¹˜ç§¯ï¼Œç„¶åé™¤ä»¥å½’ä¸€åŒ–å¸¸æ•°ï¼ˆåˆ†æ¯å’Œï¼‰
````

````python
# ç”»å›¾
plt.stem(pi_grid, posterior,
    linefmt='black',
    bottom=-1)
plt.ylim(0, 0.05) 
plt.xlabel('pi_grid')
plt.ylabel('Posterior')
sns.despine()
````

å¯ä»¥çœ‹åˆ°ï¼Œåœ¨æ­¤æ¬¡çš„æŠ½æ ·ä¸­ï¼Œ$\pi$çš„å–å€¼æ›´åŠ å¤šæ ·

````python
np.random.seed(84735)

posterior_sample = np.random.choice(pi_grid,        # pi_grid ä¸ºæŠ½æ ·çš„å€™é€‰é›†åˆï¼Œ
                                    size = 10000,   # ä» posterior åˆ†å¸ƒä¸­éšæœºæŠ½å– 10000 ä¸ªæ ·æœ¬ï¼Œä½¿ç”¨ np.random.choice()
                                    p=posterior,    # p=posterior ä¸ºæŠ½æ ·çš„æ¦‚ç‡åˆ†å¸ƒ
                                    replace=True)   # replace=True è¡¨ç¤ºæœ‰æ”¾å›æŠ½æ ·

posterior_sample = pd.DataFrame({"pi_sample": posterior_sample})    # å°†æŠ½å–çš„æ ·æœ¬å­˜å‚¨åœ¨ DataFrame ä¸­ï¼Œåˆ—åä¸º "pi_sample"

posterior_sample.value_counts(normalize=True).reset_index()        # å¯¹ posterior_sample ä¸­çš„æ ·æœ¬è¿›è¡Œè®¡æ•°ï¼Œå¹¶ä½¿ç”¨ normalize=True å°†è®¡æ•°è½¬æ¢ä¸ºç›¸å¯¹é¢‘ç‡
                                                                   # ç„¶åé€šè¿‡ reset_index() å°†ç»“æœé‡æ–°è®¾ç½®ä¸º DataFrame
````

**æŠ½æ ·ç»“æœå›¾ç¤º**  

å¯¹æ¯”ä¸€ä¸‹æŠ½æ ·å¾—åˆ°çš„åéªŒåˆ†å¸ƒå›¾ å’Œ å®é™…çš„åéªŒåˆ†å¸ƒ$Beta(11, 3)$  

ç›¸æ¯”äºä¸Šä¸€æ¬¡ï¼Œè¿™ä¸€æ¬¡çš„ç»“æœæ›´åŠ æ¥è¿‘çœŸå®çš„åéªŒåˆ†å¸ƒ

````python
x_beta = np.linspace(0,1,10000)             # ç”Ÿæˆä¸€ä¸ª 10000 ä¸ªç‚¹ï¼ŒèŒƒå›´åœ¨ [0, 1] ä¹‹é—´
y_beta = st.beta.pdf(x_beta, 11, 3)         # ç”ŸæˆBeta(11,3)

plt.plot(x_beta, y_beta, color='black')     # ç»˜åˆ¶Beta(11,3)

plt.hist(posterior_sample["pi_sample"],     # ç»˜åˆ¶10000ä¸ªæŠ½æ ·ç»“æœåˆ†å¸ƒ
         edgecolor = "black",
         color="grey",
         alpha = 0.7,
         density=True,)

sns.despine()
````

### step2: åœ¨æ¯ä¸€ä¸ª$\pi$ä¸‹ï¼Œè®¡ç®—å…ˆéªŒåˆ†å¸ƒï¼Œä¸ä¼¼ç„¶å‡½æ•°çš„ä¹˜ç§¯ï¼Œè®¡ç®—æ€»å’Œå¹¶è¿›è¡Œå½’ä¸€åŒ–

````python
# å®šä¹‰ä¼¼ç„¶å‡½æ•°ï¼Œç”±äºè¯¥ä¼¼ç„¶å‡½æ•°ä¸èƒ½é€šè¿‡è°ƒç”¨å·²æœ‰çš„å‡½æ•°è·å¾—ï¼Œå› æ­¤æˆ‘ä»¬æ ¹æ®å¾—åˆ°çš„å…¬å¼å®šä¹‰ä¸€ä¸ª
def poisson_likelihood(y):
    lambda_grid = np.linspace(0,15,501)
    
    # Calculate the Poisson likelihood for each lambda
    likelihood_poisson = np.exp(-len(y) * lambda_grid) * np.power(lambda_grid, np.sum(y)) / np.prod([np.math.factorial(val) for val in y])

    return likelihood_poisson
````

æˆ‘ä»¬å¯ä»¥ä½¿ç”¨ $\hat{R}$ é‡åŒ–ç»¼åˆé“¾å˜å¼‚æ€§å’Œé“¾å†…å˜å¼‚æ€§ä¹‹é—´çš„å…³ç³»ã€‚

$\hat{R}$ æŒ‡æ ‡ï¼ˆç§° "R-hat"ï¼‰ä¸ºå¹¶è¡Œé“¾çš„å¯è§†åŒ–æ¯”è¾ƒæä¾›äº†å…·ä½“çš„æ•°å­¦æŒ‡æ ‡ã€‚  


$$  
\hat{R} = \frac{V_{combined}}{V_{within}}

- $\hat{R}$ é€šè¿‡æ¯”è¾ƒæ‰€æœ‰é“¾ä¸Šé‡‡æ ·å€¼çš„å˜å¼‚æ€§å’Œæ¯ä¸ªå•ç‹¬é“¾ä¸Šçš„å˜å¼‚æ€§æ¥è§£å†³ä¸€è‡´æ€§é—®é¢˜ã€‚

**ä»£ç æ¼”ç¤º**  

ä½¿ç”¨ pymc å»ºæ¨¡çš„æ¨¡å‹é»˜è®¤ä½¿ç”¨ MCMC æ–¹æ³•è¿›è¡Œå‚æ•°ä¼°è®¡ã€‚  

å¹¶ä¸”æˆ‘ä»¬å¯ä»¥å¾ˆæ–¹ä¾¿çš„ä½¿ç”¨ `arviz` å¯¹MCMCè¿›è¡Œå¯è§†åŒ–å’ŒåŸºäºæ•°å­¦åŒ–æ ‡å‡†çš„è¯Šæ–­

é¦–å…ˆï¼Œæˆ‘ä»¬å»ºç«‹ä¸€ä¸ªç®€å•çš„ beta æ¨¡å‹ï¼Œå¹¶å¯¹è¯¥æ¨¡å‹ä½¿ç”¨MCMCé‡‡æ ·æ–¹æ³•  
- è®¾ç½®é‡‡æ ·æ•° draws ä¸º 5000  
- è®¾ç½® MCMC é“¾çš„æ•°é‡ä¸º 4ï¼Œå’Œä¸Šé¢çš„ç¤ºä¾‹ä¸€æ ·ã€‚  
- è®¾ç½® MCMC è°ƒè¯•çš„æ•°é‡ tune ä¸º0ï¼Œè€Œé»˜è®¤ tune ä¸º 1000ã€‚  
    - è¿™é‡Œæš‚æ—¶ç†è§£ tune ä¸ºæå‰è¿›è¡Œ MCMC é‡‡æ · 1000æ¬¡ï¼Œç„¶åä¸¢å¼ƒæ‰è¿™äº›é‡‡æ ·ã€‚  
    - æ ¹æ® MCMC çš„æ€§è´¨ï¼Œé“¾ä¸­è¶Šæ—©çš„é‡‡æ ·è¶Šä¸ç¨³å®šã€‚å› æ­¤ï¼Œä¸¢æ‰è¿™äº›é‡‡æ ·å¯ä»¥ä¿è¯åç»­çš„é‡‡æ ·æ›´åŠ ç¨³å®šã€‚  
    - åç»­è¯¾ç¨‹ä¼šè¯¦ç»†ä»‹ç»è¯¥å‚æ•°çš„æ„ä¹‰ã€‚

````python
import pymc as pm
import arviz as az

# è®¾ç½®æ•°æ®
Y = 9
# åˆ›å»ºPyMCæ¨¡å‹
bb_model = pm.Model()

with bb_model:
    pi = pm.Beta('pi', alpha=2, beta=2)
    likelihood = pm.Binomial('likelihood', n=10, p=pi, observed=Y)

#é‡‡æ ·è¿‡ç¨‹ä»åœ¨è¯¥å®¹å™¨ä¸­è¿›è¡Œ
with bb_model:
    trace = pm.sample(draws=5000,                   # è®¾ç½®MCMCé‡‡æ ·æ¬¡æ•°
                      tune=0,                       # è®¾ç½®è°ƒè¯•å’Œéœ€è¦ä¸¢å¼ƒçš„é‡‡æ ·æ•°
                      chains=4,                     # è®¾ç½®4æ¡é“¾
                      random_seed=84735)
````

````python
az.plot_trace(trace, legend = True)
````

````python
az.plot_autocorr(trace, max_lag=20, combined=True)
````
