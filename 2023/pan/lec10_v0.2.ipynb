### Median absolute error (MAE)

ç»å¯¹è¯¯å·®çš„ä¸­ä½æ•°ï¼Œmedian absolute error (MAE)è¡¡é‡äº†é¢„æµ‹è§‚æµ‹å€¼$Y_i$ä¸åéªŒé¢„æµ‹å‡å€¼ä¹‹é—´$Y_i'$çš„å·®å¼‚ã€‚

$$
\text{MAE} = median\left(|Y_i - \hat{Y}_i|\right)
$$

- å‡è®¾ Y1 Y2â€¦ï¼Œ Ynè¡¨ç¤ºnä¸ªè§‚å¯Ÿç»“æœã€‚
- æ¯ä¸ª $Y_i$ éƒ½æœ‰å¯¹åº”çš„åéªŒé¢„æµ‹æ¨¡å‹ï¼Œå…¶å‡å€¼ä¸º $Y_i'$ï¼Œå…¶æ ‡å‡†å·®ä¸º $sd_i$ã€‚

ç»“åˆæˆ‘ä»¬ä¹‹å‰çš„ä¾‹å­æ¥ç†è§£ç»å¯¹è¯¯å·®çš„ä¸­ä½æ•°ï¼š
- é¦–å…ˆï¼Œæˆ‘ä»¬ä»¥å‹åŠ› $X_i=45$ ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹ä¸ºä¾‹å­ã€‚
- ä¹‹åï¼Œæˆ‘ä»¬è®¡ç®—åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼å’Œè§‚æµ‹å€¼ä¹‹é—´çš„è¯¯å·®ã€‚
- å†æ¬¡ï¼Œæˆ‘ä»¬è®¡ç®—æ‰€æœ‰è§‚æµ‹å˜é‡ä¸‹åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼å’Œè§‚æµ‹å€¼ä¹‹é—´çš„è¯¯å·®ã€‚
- æœ€åï¼Œæˆ‘ä»¬è®¡ç®—æ‰€æœ‰è¯¯å·®çš„ä¸­ä½æ•°ã€‚

![Image Name](https://cdn.kesci.com/upload/s3y4rmsym1.png?imageView2/0/w/960/h/960)

**é¦–å…ˆï¼Œæˆ‘ä»¬è€ƒè™‘å‹åŠ› $X_i=45$ ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹**

1. å’Œä¹‹å‰ä¸€æ ·ï¼Œæˆ‘ä»¬é¦–å…ˆè¯»å–æ•°æ® Data_Sum_HPP_Multi_Site_Shareï¼Œé€‰æ‹©"Tsinghua"ç«™ç‚¹çš„æ•°æ®ä½œä¸ºç¤ºä¾‹ã€‚
2. ä¹‹åï¼Œä½¿ç”¨ pymc å¯¹å‹åŠ›å˜é‡`stress` å’Œè‡ªæˆ‘æ§åˆ¶å˜é‡ `scontrol` å»ºç«‹å›å½’æ¨¡å‹ã€‚
3. åœ¨æ¨¡å‹æ‹Ÿåˆå(ç•¥è¿‡æ¨¡å‹è¯Šæ–­ç¯å¢ƒ)ï¼Œç”Ÿæˆæ¨¡å‹çš„åéªŒé¢„æµ‹ã€‚
4. æœ€åï¼Œæˆ‘ä»¬è®¡ç®—å‹åŠ› $X_i=45$æ¡ä»¶ä¸‹ï¼ŒåéªŒé¢„æµ‹çš„å‡å€¼å’Œè§‚æµ‹å€¼ä¹‹é—´çš„è¯¯å·®ã€‚

```python
# å¯¼å…¥ pymc æ¨¡å‹åŒ…ï¼Œå’Œ arviz ç­‰åˆ†æå·¥å…· 
import pymc as pm
import arviz as az
import seaborn as sns
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd

# å¿½ç•¥ä¸å¿…è¦çš„è­¦å‘Š
import warnings
warnings.filterwarnings("ignore")
```

```python
# é€šè¿‡ pd.read_csv åŠ è½½æ•°æ® Data_Sum_HPP_Multi_Site_Share.csv
df_raw = pd.read_csv('../lin/lec10/Data_Sum_HPP_Multi_Site_Share.csv')

# ç­›é€‰ç«™ç‚¹ä¸º"Tsinghua"çš„æ•°æ®
df = df_raw[df_raw["Site"] == "Tsinghua"]

# é€‰æ‹©éœ€è¦çš„ä¸¤åˆ—
df = df[["stress","scontrol"]]

#è®¾ç½®ç´¢å¼•
df["index"] = range(len(df))
df = df.set_index("index")
```

```python
# å¯¹å‹åŠ›å˜é‡`stress` å’Œè‡ªæˆ‘æ§åˆ¶å˜é‡ `scontrol` å»ºç«‹å›å½’æ¨¡å‹
with pm.Model() as linear_model:

    beta_0 = pm.Normal("beta_0", mu=70, sigma=4)          #å®šä¹‰beta_0          
    beta_1 = pm.Normal("beta_1", mu=0, sigma=0.8)         #å®šä¹‰beta_1
    sigma = pm.Exponential("sigma", 0.6)                  #å®šä¹‰sigma

    #xæ˜¯è‡ªå˜é‡å‹åŠ›æ°´å¹³
    x = pm.MutableData("x",df.stress,dims="obs_id")       
    #å®šä¹‰muï¼Œè®²è‡ªå˜é‡ä¸å…ˆéªŒç»“åˆ              
    mu = pm.Deterministic("mu", beta_0 + beta_1 * x, dims="obs_id")    
    #å®šä¹‰ä¼¼ç„¶ï¼šé¢„æµ‹å€¼yç¬¦åˆN(mu, sigma)åˆ†å¸ƒ, observed ä¸ºä¼ å…¥å®é™…æ•°æ®y è‡ªæˆ‘æ§åˆ¶æ°´å¹³
    likelihood = pm.Normal("y_est", mu=mu, sigma=sigma, observed=df.scontrol, dims="obs_id")   
```

```python
# æ¨¡å‹æ‹Ÿåˆ 
#===========================
#     æ³¨æ„ï¼ï¼ï¼ä»¥ä¸‹ä»£ç å¯èƒ½éœ€è¦è¿è¡Œ1-2åˆ†é’Ÿå·¦å³
#===========================
with linear_model:
    trace = pm.sample(draws=5000,                   # ä½¿ç”¨mcmcæ–¹æ³•è¿›è¡Œé‡‡æ ·ï¼Œdrawsä¸ºé‡‡æ ·æ¬¡æ•°
                      tune=1000,                    # tuneä¸ºè°ƒæ•´é‡‡æ ·ç­–ç•¥çš„æ¬¡æ•°ï¼Œå¯ä»¥å†³å®šè¿™äº›ç»“æœæ˜¯å¦è¦è¢«ä¿ç•™
                      chains=4,                     # é“¾æ•°
                      discard_tuned_samples= True,  # tuneçš„ç»“æœå°†åœ¨é‡‡æ ·ç»“æŸåè¢«ä¸¢å¼ƒ
                      random_seed=84735)
```

- é€šè¿‡ `pm.set_data({"x": np.array([45])})` è®¾ç½®æ¨¡å‹ç”Ÿæˆx=40æ¡ä»¶ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹
- ç”±äºè‡ªå®šä¹‰çš„æ•°æ®å¯èƒ½ä¸ºæ ·æœ¬å¤–æ•°æ®ï¼Œå› æ­¤éœ€è¦è®¾ç½® `predictions=True`è¿›è¡Œæ ·æœ¬å¤–é¢„æµ‹ã€‚
- æ ·æœ¬å¤–é¢„æµ‹ç»“æœå‚¨å­˜åœ¨ `ppc.predictions` ä¸­ã€‚è€Œ `ppc.predictions_constant_data`æ˜¯æˆ‘ä»¬è®¾ç½®çš„x=40çš„æ•°æ®ã€‚

```python
# ç”Ÿæˆæ¨¡å‹çš„åéªŒé¢„æµ‹
with linear_model:

    pm.set_data({"x": np.array([45])})   # æ³¨æ„ï¼šä»…ç”Ÿæˆx=45æ¡ä»¶ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹

    ppc = pm.sample_posterior_predictive(
        trace, 
        var_names=["y_est"],
        predictions=True,
        random_seed=84735
    )
```

**è®¡ç®—æ‰€æœ‰è¯¯å·®çš„ä¸­ä½æ•°(MAE)**

$$
{\mathsf{MAE}}={\mathsf{median}}|Y_{i}-Y_{i}^{\prime}|.
$$

- MAE ä¸ºæ‰€æœ‰é¢„æµ‹è¯¯å·®Y-Y'çš„ä¸­ä½æ•°ã€‚

æœ€åï¼Œæˆ‘ä»¬è®¡ç®—æ‰€æœ‰è§‚æµ‹å€¼ä¸åéªŒé¢„æµ‹æ¨¡å‹å‡å€¼ä¹‹é—´çš„è¯¯å·®ã€‚

- é¦–å…ˆï¼Œæˆ‘ä»¬å°†ä¹‹å‰è®¡ç®—åéªŒé¢„æµ‹è¯¯å·®(prediction error)çš„æ–¹æ³•å°è£…ä¸ºå‡½æ•°ï¼Œæ–¹ä¾¿åç»­è®¡ç®—ã€‚
- ç„¶åï¼Œä½¿ç”¨å¾ªç¯æ–¹æ³•ï¼Œè®¡ç®—æ¯ç§Xè°ƒèŠ‚ä¸‹çš„é¢„æµ‹è¯¯å·®ã€‚
- æœ€åï¼Œåˆå¹¶æ‰€æœ‰è¯¯å·®ï¼Œå¹¶è®¡ç®—è¿™äº›è¯¯å·®çš„ä¸­ä½æ•°ã€‚

```python
def prediction_error(X=45):
  # æå–å¯¹åº”Xå€¼ä¸‹çš„è§‚æµ‹å€¼Y
  true_ys = df[df.stress == X]["scontrol"].values
  # æå–å¯¹åº”Xå€¼ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼
  pre_y_mean_i = pre_y_mean[pre_y_mean.stress == X]["scontrol_ppc_mean"].values
  # è®¡ç®—é¢„æµ‹è¯¯å·®
  pre_error = true_ys - pre_y_mean_i
  return pre_error

# è®¡ç®—Xä¸º45æ—¶çš„é¢„æµ‹è¯¯å·®
print("prediction error:\n", prediction_error(X=45))
```

```python
# å¯¼å…¥ median å‡½æ•°ï¼Œç”¨äºè®¡ç®—ä¸­ä½æ•°
from statistics import median
# å®šä¹‰ç©ºåˆ—è¡¨ï¼Œç”¨äºåˆå¹¶æ‰€æœ‰é¢„æµ‹è¯¯å·®
pre_errors = []


# ä½¿ç”¨å¾ªç¯æ–¹æ³•ï¼Œè®¡ç®—æ¯ç§Xè°ƒèŠ‚ä¸‹çš„é¢„æµ‹è¯¯å·®ã€‚
for x in x_stress:                      # x_stress æˆ‘ä»¬è·å¾—çš„Xçš„32ç§å¯èƒ½çš„å–å€¼
  pre_errors.append(prediction_error(X=x))

# æœ€åï¼Œè®¡ç®—é¢„æµ‹è¯¯å·®çš„ä¸­ä½æ•°ã€‚
pre_errors = np.concatenate(pre_errors)  # å°†é¢„æµ‹è¯¯å·®åˆå¹¶ä¸ºnumpy array
MAE = median(pre_errors)
print(f"è®¡ç®—å¾—åˆ°çš„MAE: {MAE:.2f}")
```

```python
# å¯è§†åŒ–é¢„æµ‹è¯¯å·®å’ŒMAE
az.plot_posterior(pre_errors, hdi_prob=0.95, point_estimate="median", ref_val=MAE)
```

* åœ¨å½“å‰æ•°æ®é›†ä¸­ï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®— 32 ç§Xå–å€¼ä¸‹å¯¹åº”çš„åéªŒé¢„æµ‹æ¨¡å‹çš„å‡å€¼ã€‚
  * ä»£å…¥`20000`ä¸ªå‚æ•°å¯¹ï¼Œæ¯ä¸ªè‡ªå˜é‡éƒ½æ‹¥æœ‰å¯¹åº”çš„20000ä¸ªåéªŒé¢„æµ‹å€¼
* è¿›è€Œï¼Œæˆ‘ä»¬å¯ä»¥è®¡ç®—å‡º32ä¸ªå¯ä¿¡åŒºé—´ï¼Œå¹¶è§‚å¯ŸçœŸå®å€¼è½åœ¨å¯ä¿¡åŒºé—´çš„æƒ…å†µã€‚


```python
# è·å–xå‹åŠ›æ•°æ®çš„32ç§å¯èƒ½çš„å€¼
x_stress = df.stress.unique()

# ç”Ÿæˆ32ç§Xå–å€¼ä¸‹å¯¹åº”çš„åéªŒé¢„æµ‹æ¨¡å‹
with linear_model:

    pm.set_data({"x": x_stress})   # è®¾ç½®ç”Ÿæˆ32ç§Xæ¡ä»¶ä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹

    ppc = pm.sample_posterior_predictive(
        trace, 
        var_names=["y_est"],
        predictions=True,
        extend_inferencedata=True,
        random_seed=84735
    )
```

æˆ‘ä»¬å°†ä¹‹å‰è®¡ç®—å•ä¸ªXå¯¹åº”ä¸‹åéªŒé¢„æµ‹æ¨¡å‹HDIçš„æ–¹å¼æ‰“åŒ…ä¸ºå‡½æ•°ï¼š

- é¦–å…ˆï¼Œå®šä¹‰å‡½æ•°ï¼Œè®¡ç®—å¯¹åº”Xä¸‹çš„åéªŒé¢„æµ‹æ¨¡å‹hdièŒƒå›´
- å…¶æ¬¡ï¼Œåˆ¤æ–­è§‚æµ‹å€¼æ˜¯å¦è¶…è¿‡ hdi è¾¹ç•Œ
- æœ€åï¼Œè®°å½•å¹¶è¿”å›è¶…è¿‡ hdi è¾¹ç•Œçš„è§‚æµ‹å€¼çš„æ•°é‡

```python
# å®šä¹‰å‡½æ•°ï¼Œè®¡ç®—å¯¹åº”Xä¸‹çš„è§‚æµ‹å€¼è¶…å‡ºåéªŒé¢„æµ‹æ¨¡å‹hdièŒƒå›´çš„æ•°é‡
def count_outlier(i,hdi=0.95):
  # æå–ç¬¬iä¸ªXçš„å€¼
  X = ppc.predictions_constant_data.x.isel(obs_id=i).values
  # æå–ç¬¬iä¸ªXå€¼å¯¹åº”çš„åéªŒé¢„æµ‹å€¼
  pre_ys = ppc.predictions["y_est"].stack(sample=("chain", "draw")).isel(obs_id=0).values
  # æå–å¯¹åº”Xå€¼ä¸‹çš„è§‚æµ‹å€¼Y
  true_ys = df[df.stress == X]["scontrol"]
  
  # è®¡ç®—å¯¹åº”åéªŒé¢„æµ‹çš„hdiçš„ä¸Šä¸‹é™
  lower,upper = az.hdi(pre_ys, hdi_prob=hdi)
  
  # åˆ¤æ–­æ˜¯å¦è¶…è¿‡ hdi è¾¹ç•Œ
  outlier = (true_ys > upper) | (true_ys < lower)

  # è®¡ç®—è¶…å‡ºè¾¹ç•Œçš„è§‚æµ‹å€¼çš„æ•°é‡
  num_outlier = sum(outlier)
  return num_outlier

# è®¡ç®—Xä¸º45æ—¶çš„é¢„æµ‹è¯¯å·®
i = 5
print(f"Xä¸º{x_stress[i]}ä¸‹ï¼Œè¶…è¿‡åéªŒé¢„æµ‹èŒƒå›´è§‚æµ‹å€¼çš„æ•°é‡:", count_outlier(i))
```

æ¥ç€ï¼Œæˆ‘ä»¬å¯ä»¥é€šè¿‡å¾ªç¯çš„æ–¹å¼è®¡ç®—æ¯ä¸ªXå¯¹åº”ä¸‹è¶…è¿‡åéªŒé¢„æµ‹èŒƒå›´è§‚æµ‹å€¼çš„æ•°é‡ã€‚

```python
# å®šä¹‰æ‰€æœ‰è¶…è¿‡åéªŒé¢„æµ‹èŒƒå›´è§‚æµ‹å€¼çš„æ•°é‡
total_number = 0

# ä½¿ç”¨å¾ªç¯æ–¹æ³•ï¼Œå¾ªç¯32ç§Xè°ƒèŠ‚ä¸‹çš„é¢„æµ‹è¯¯å·®ã€‚
for i in range(32):  
  total_number += count_outlier(i)

print(f"æ‰€æœ‰è¶…è¿‡åéªŒé¢„æµ‹èŒƒå›´è§‚æµ‹å€¼çš„æ•°é‡: {total_number}")
```
