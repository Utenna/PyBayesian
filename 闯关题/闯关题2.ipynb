## STEP1ï¼šæ•°æ®ç¼–ç   

1. é—®é¢˜q1ï¼šå°†æ•°æ®ä¸­`Sex`åˆ—çš„å€¼è¿›è¡Œè½¬æ¢ï¼Œå…¶ä¸­0è¡¨ç¤º`female`ï¼Œ1è¡¨ç¤º`male`ã€‚  
      - æç¤ºï¼šå¯ä»¥ä½¿ç”¨ `df.map()`ã€`df.replace()` ç­‰æ–¹æ³•ï¼Œå°½é‡ä¸è¦ä½¿ç”¨ `if else` ç­‰æ–¹æ³•ã€‚

```python
# é—®é¢˜q1ï¼šå°†æ•°æ®ä¸­`Sex`åˆ—çš„å€¼è¿›è¡Œè½¬æ¢ï¼Œå…¶ä¸­0è¡¨ç¤º`female`ï¼Œ1è¡¨ç¤º`male`
# æç¤ºï¼šdf.map() æ˜¯å¯¹ Seriesï¼ˆå•åˆ—ï¼‰è¿›è¡Œå…ƒç´ çº§æ˜ å°„çš„éå¸¸ç®€å•çš„å‡½æ•°ã€‚æˆ‘ä»¬å¯ä»¥ä¼ å…¥ä¸€ä¸ªå­—å…¸ï¼Œå°†æ•°å€¼æ˜ å°„åˆ°å¯¹åº”çš„å­—ç¬¦ä¸²ã€‚
# æç¤ºï¼šdf.replace() åˆ™å¯ä»¥ç”¨äºæ›¿æ¢ DataFrame ä¸­çš„å€¼ï¼Œæ—¢å¯ä»¥æ›¿æ¢åˆ—ä¸­çš„å€¼ï¼Œä¹Ÿå¯ä»¥è¿›è¡Œå¤šåˆ—çš„æ›¿æ¢ã€‚
df["Sex"] = df['Sex'].map({0:'female',1:'male'})

# é—®é¢˜q2ï¼šå¯¹äºæ•°æ®ä¸­çš„ `EverythingIsEffort` åˆ—è¿›è¡Œæ–¹å‘è®¡åˆ†ç¼–ç (å€¼ 4é‡æ–°ç¼–ç ä¸º0ï¼Œå°†3é‡æ–°ç¼–ç ä¸º1ï¼Œå°†2é‡æ–°ç¼–ç ä¸º2ï¼Œå°†1é‡æ–°ç¼–ç ä¸º3ï¼Œå°†0é‡æ–°ç¼–ç ä¸º4)ã€‚
df["EverythingIsEffort_recode"] = df['EverythingIsEffort'].map({})
```
